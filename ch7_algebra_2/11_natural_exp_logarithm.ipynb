```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x1 = np.linspace(-3, 3, n)
x2 = np.linspace(0.01, 10, n)
y1 = np.exp(x1)
y2 = np.log(x2)
```

```python
plt.plot(x1, y1)
plt.xlabel('x')
plt.ylabel('$e^x$')
plt.show()

plt.plot(x2, y2)
plt.xlabel('x')
plt.ylabel('$log(x)$')
plt.show()
```

```python
# Exercise
x = np.linspace(0.0001, 10, 12)
y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x, y1, label='$log(exp^x)$')
plt.plot(x, y2, 'ro', label='$e^{log(x)}$')

plt.axis('square')
plt.legend()
plt.show()
```
