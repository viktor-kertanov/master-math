```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
import sympy as sym
```

```python
cols = 1000
rows = 100

t = np.linspace(-3, 3, cols)
h = np.linspace(0.01, 1, rows)

G = np.zeros([rows, cols])
print(G.shape)
```

```python
for h_idx, h_i in enumerate(h):
    for t_idx, t_i in enumerate(t):
        # G[h_idx, t_idx] = np.exp((-4*np.log(2)*t_i**2) / (h_i**2))
        G[h_idx, :] = np.exp((-4*np.log(2)*t_i**2) / (h_i**2))
```

```python
plt.pcolormesh(t, h, G)

plt.xlabel('t')
plt.ylabel('h')
plt.show()
```

```python
A = np.zeros([rows, cols])
for h_idx, h_i in enumerate(h):
    A[h_idx, :] = np.exp((-4*np.log(2)*t**2) / (h_i**2))
```

```python
plt.pcolormesh(A)

plt.show()
```
