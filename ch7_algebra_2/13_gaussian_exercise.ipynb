```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = 1001
rows = 1001

t = np.linspace(-3, 3, cols)
h = np.linspace(0.001, 1, rows)

G = np.zeros([rows, cols])
```

```python
for r in range(rows):
    for c in range(cols):
        G[r, c] = np.exp((-4*np.log(2)*t[c]**2) / (h[r]**2))
```

```python
plt.plot(G)

plt.xlabel('t')
plt.ylabel('h')
plt.show()
```
