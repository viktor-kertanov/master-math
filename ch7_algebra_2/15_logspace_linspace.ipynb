```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
import sympy as sym
```

Linearly-spaced numbers is a simple concept: we divide the "line" between the first and the second number into n equal segments, each of the segment is similar in length/size

```python
music_octave_linear = np.linspace(440, 880, 8)

display(Math(sym.latex(sym.sympify(music_octave_linear))))
```

Below in the cell it is seen that np.logspace() works not like linspace. The first number is $10^1$, the second number is also a power of ten: $10^2$. And these 2 numbers 10 and 100 are splitted into unevenely=log-spaced intervals

```python
np.logspace(1, 2, 10)
```

So, if we go back to the music-scale theme, then we should do the following.
The only way to get proper, standard music frequencies, was to take a full scale with half tones (12 half-tones and 13th is the same note but in the next octave) and divide it into 13 equal segments, log-spaced.

```python
note_a_1 = np.log10(220) # note a in the first octave, Hz; prepared for np.logspace()
note_a_2 = np.log10(440) # note a in the second octave, Hz; prepared for np.logspace()

music_scale_log = np.logspace(note_a_1, note_a_2, 13)
display(Math(sym.latex(sym.sympify(music_scale_log))))
```

```python
# Exercise: we should show that as boundaries become smaller and smaller linear functions becomes better and better in approximation of logarithmic function

lin_fx = np.linspace(0, 100, 100)
log_fx = np.logspace(0, 2, 100)

plt.plot(lin_fx, label='linear')
plt.plot(log_fx, label='logspace')
plt.axis('square')
plt.axis([-5,105,-5,105])
plt.legend()
plt.show()
```

It's a very importand fact that linear functions become really good approximations of non-linear functions as the distance gets smaller.

```python
a = 2
b = 100
n = 50

for delta in range(0, 45, 5):
    li = np.linspace(a+delta, b-delta, n)
    lo = np.logspace(np.log10(a+delta), np.log10(b-delta), n)

    plt.plot(li, lo, label='log')

li = np.linspace(a, b, n)
plt.plot(li, li, label='linear', color='k')
plt.axis('square')
plt.show()
```
