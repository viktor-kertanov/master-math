```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Yt = Xt+1- Xt 

x = [1, 3, 5, 8, 99, -4]

x_diff = np.diff(x, n=1)
x_diff_2 = np.diff(x, n=2) # second iteration of the same discrete derivative function
x_diff_3 = np.diff(x, n=3) # third iteration of the same discrete derivative function ; the number of point is getting less by one with each iteration

print(x_diff)
print(x_diff_2)
print(x_diff_3)
```

```python
# Exercise
n=100
x = np.linspace(-4, 4, n)
y = x**2
```

```python
diff_x = np.diff(x)
diff_f = np.diff(y)
dx_df = diff_f/diff_x
```

```python
plt.plot(x, y, label=f'$f(x)=x^2$')
plt.plot(x[:-1], dx_df, label='$dx/df$')

plt.grid()
plt.ylim([-1,2])
plt.xlim([-2,2])
plt.legend()
plt.show()
```
