```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = complex(2, 3)

plt.plot(z.real, z.imag, 'ro')
plt.plot([0, z.real], [0, z.imag], 'r')

plt.xlabel('real')
plt.ylabel('imaginary')
plt.axis([-4, 4, -4, 4])
plt.grid()
plt.show()
```

```python
#exercise

z1 = complex(-3, 1)
z2 = complex(-1, 1)
sum_z = z1+z2
prod_z = z1*z2

lines = [z1, z2, sum_z, prod_z]
labels = ["z1", "z2", "z1 + z2", "z1 * z2"]

for idx, line in enumerate(lines, start=1):
    print(f'{idx}. Real: {line.real}, Imag: {line.imag}')
    plt.plot([0, line.real], [0, line.imag], label=f'{labels[idx-1]}')

plt.axis([-5,5,-5,5])
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.legend()
plt.grid()
plt.show()
```
