```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math 
```

```python
t = np.linspace(-3, 3, 100)
h_init = 0.5
for h_factor in range(1, 10):
    y = np.exp((-4*np.log(2)*t**2) / ((h_init*h_factor)**2))
    plt.plot(t, y, label=f'h={h_init*h_factor}')

plt.legend()
plt.show()
```

```python
h = 1
t = np.linspace(-3, 3, 1001)
g = np.exp((-4*np.log(2) * t**2) / (h**2))

plt.plot(t, g)

plt.xlabel('t')
plt.ylabel('y')
plt.title('The Gaussian with Full Width at Half Maximum (FWHM) at %s' % h)

y_value_to_find = 0.5
x_value = h  * np.sqrt(np.log(y_value_to_find)/(-4*np.log(2))) #be careful with PEMDAS

x1 = -x_value
x2 = x_value
print(f'x1 and x2: {x1}, {x2}')

plt.plot(x1, y_value_to_find, 'ro')
plt.plot(x2, y_value_to_find, 'ro')

plt.show()
```
