# Roots of Unity

```python
import sympy as sym
from IPython.display import display, Math
import numpy as np
import matplotlib.pyplot as plt 
```

$
\text{Roots of unity means that theres a complex number z, such that: } \newline z^n = 1,\quad z \in \mathbb{C}.
\newline \text{The solution (roots) is for each k: } z = e^{2 \pi i k / n}, \quad k\in [0, 1, ..., n-1]
\newline
\text{If n = 5, it meas there are 5 roots}
$

```python
n=5

for k in range(5):
    z1 = np.exp((2* np.pi * 1j *k )/n) #this won't give us exactly one, only a close estimation
    z = sym.exp((2*sym.pi* sym.I * k) / n)
    display(Math('(%s)^{%s} \\quad \\Rightarrow \\quad %s' % (sym.latex(z), n, sym.latex(z**n))))
```

#### We will plot this complex roots of unity

The interesting thing is that the roots of unity give us an evenly splitted circle. If there are 5 roots of unity, than we'll have a unit circle splitted into 5 equal parts.

```python
n=11

for k in range(n): 
    z = sym.exp(2 * sym.pi * sym.I * k / n)
    plt.plot([0, sym.re(z)], [0, sym.im(z)])

# let's draw a circle
circle_n = 1000
angles = np.linspace(0, 2*np.pi, circle_n)
x = np.cos(angles)
y = np.sin(angles)
plt.plot(x, y, color='gray')

plt.axis('square')
plt.show()
```

```python
n=250
g_col = np.linspace(0, .9, n)

for k in range(n): 
    z = np.exp(2 * np.pi * 1j* k/n)*k
    plt.plot([0, np.real(z)], [0, np.imag(z)], color=(g_col[k], g_col[k], g_col[k]))

plt.axis('square')
plt.axis('off')
plt.savefig(f'root_of_unity_ex.svg')
plt.show()
```
