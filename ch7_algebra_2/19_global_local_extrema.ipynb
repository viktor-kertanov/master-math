```python
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
import sympy as sym
from scipy.signal import find_peaks
```

# How to find global and local extrema of a function: maxima, minima ?

We need to be able to find not only the maxima/minima points, but also identify their location - where in the data set they are located.

```python
lst = [-4, 5.5, 1, 2]

max_val = np.max(lst)
print(max_val)

max_val_idx = np.argmax(lst)
print(max_val_idx)
```

Let's create a function and then find maximum/minimum values

```python
x = np.linspace(0, 2*np.pi, 20)
fx = -(np.cos(x) + np.sqrt(x))


plt.plot(x, fx, 'bo-')
plt.plot(x[np.argmax(fx)], np.max(fx), 'rs') # with red circle we show the maximum of the function 

plt.show()
```

```python
display(Math(sym.latex(sym.sympify(fx))))
max_val = np.max(fx)
max_val_idx = np.argmax(fx)

print(f'Maximum value is: {max_val}. Idx in the dataset: {max_val_idx}')
```

```python
# the next function to find min
x = np.linspace(-2, 2, 51)

fx_poly = x**3 + x**4 

plt.plot(x, fx_poly, 'bo-')
plt.plot(x[np.argmin(fx_poly)], np.min(fx_poly), 'rs')

plt.xlim([-1, 0.5]) #zooming in function to see a clear minimum
plt.ylim([-.2, .2])
plt.show()
```

```python
# Exercise: finding not only a global max/min but local peaks, using find_peaks module

x = np.linspace(0, 12*np.pi, 1000)
fx = -np.cos(x) - np.sqrt(x)

plt.plot(x, fx)

# now let's take care of the peaks
peaks = find_peaks(fx)

#possible, but commented out as vectors are better
# for peak in peaks[0]:
#     print(f'Our peak idx in dataset: {peak}')
#     plt.plot(x[peak], fx[peak], 'ro')

plt.plot(x[peaks[0]], fx[peaks[0]], 'go')

plt.show()
```
