```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha_angles = np.linspace(0, 2*np.pi, 360)
x = np.cos(alpha_angles)
y = np.sin(alpha_angles)
```

```python
n_random = 4
random_angles = np.random.choice(alpha_angles, size=n_random)

max_scl = 1.5

plt.plot(x, y, 'k')

plt.plot([-max_scl, max_scl], [0,0], 'k--', alpha=0.3)
plt.plot([0, 0], [-max_scl, max_scl], 'k--', alpha=0.3)

#random plot
for random_angle in random_angles:
    plt.plot([0, np.cos(random_angle)], [0, np.sin(random_angle)])

#beautify
plt.axis('square')
plt.axis([-max_scl, max_scl, -max_scl, max_scl])
plt.xlabel('cosine')
plt.ylabel('sine')

plt.show()
```

```python
# exercise
n = 50
angles = np.linspace(-3, 1.83, n)
k = np.exp(angles)
for angle in k:
    plt.plot([0, np.cos(angle)], [0, np.sin(angle)])

plt.axis('square')
plt.axis('off')
plt.show()
```

```python
n = 100
lines = np.linspace(-3, 3, n)
k = 1.4**lines
for line in k:
    plt.plot([np.random.normal(-2, 0.05, 1), np.random.normal(2, 0.05, 1)], [line, line], linewidth=0.5, color='k')

plt.axis('on')
plt.savefig('exp_lines.svg')
plt.show()
```

```python
# exercise
n = np.random.randint(7, 100)
angles = np.linspace(0, 2*np.pi, n)
k = np.exp(angles)
for angle in k:
    plt.plot([0, np.cos(angle)], [0, np.sin(angle)], color='k', linewidth=0.1)

plt.axis('square')
plt.axis('off')
plt.savefig(f'exp_unit_circle_{n}.svg')
plt.show()
```
