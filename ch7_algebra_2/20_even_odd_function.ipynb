# Even and odd functions

$\text{If }f(x) = f(-x) \text{, than the } f(x) \text{ is even. Example: parabola} \newline$
$\text{If }f(x) = -f(x) \text{, than the } f(x) \text{ is odd. Example: } x^{3} \newline$ 

$\text{Not all functions even or odd: many functions neither even nor odd.}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 20 )

fx_even = x**2
fx_even_neg = (-x)**2

plt.plot(x, fx_even)
plt.plot(x, fx_even_neg, 'ro')

plt.show()
```

```python
x = np.linspace(-5, 5, 40)
fx_odd = x**3
fx_odd_neg = (-x)**3
fx_odd_neg_check = -x**3

plt.plot(x, fx_odd, label='$x^3$')
plt.plot(x, fx_odd_neg, 'ro', label='$(-x)^3$')
plt.plot(x, fx_odd_neg_check, 'k--', label='$-x^3$')

plt.legend()
plt.show()
```

```python
# Now we'll have to determine if sin(x), cos(x) are even/odd 

x = np.linspace(-4*np.pi, 4*np.pi, 1000)
sin_x = np.sin(x)
sin_x_negative = np.sin(-x)
neg_sin_x = -np.sin(x)

plt.plot(x, sin_x, label='$sin(x)')
plt.plot(x, sin_x_negative, 'r', label='$sin(-x)$')
plt.plot(x, neg_sin_x,'g--', label='$-sin(x)', )

plt.axis([-13, 13, -1.75, 1.75])
plt.legend()
plt.show()
```

```python
x = np.linspace(-4*np.pi, 4*np.pi, 1000)

cos_x = np.cos(x)
sin_x_negative = np.cos(-x)
neg_sin_x = -np.cos(x)

plt.plot(x, sin_x, 'b', label='$cos(x)')
plt.plot(x, sin_x_negative, 'g--', label='$cos(-x)$')
plt.plot(x, neg_sin_x,'k--', label='$-cos(x)', )

plt.axis([-13, 13, -1.75, 1.75])
plt.legend()
plt.show()
```

ANSWER: So sin(x) is an odd function and cos(x) is an even function.
