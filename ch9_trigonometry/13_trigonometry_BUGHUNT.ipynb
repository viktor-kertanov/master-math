```python
# plot a series of random numbers
s = np.random.randn(1000,2)
plt.plot(s, 'rs', alpha=0.1)
plt.show()
```

```python
# create and image a matrix of random integers between (and including) 3 and 20

mat = np.random.randint(3,21,[30,20])

plt.imshow(mat)
plt.colorbar()
plt.show()

print(np.max(mat), np.min(mat))
```

```python
# create 100 random phase angles [0,2pi] and show unit vectors with those angles

n = 100

randphases = np.random.rand(n)*2*np.pi

for i in range(n):
    plt.polar([0, randphases[i]],[0, 1], 'o-', color=np.random.rand(3), markersize=20, alpha=.3)
    
plt.show()
```

```python
# create an outwards spiral using phase angles and amplitudes

n = 10000
a = np.linspace(0, 100, n)
p = np.linspace(0, 200*np.pi, n)

plt.polar(p, a, 'k', linewidth=0.5)

plt.axis('off')

plt.savefig('simple_spiral.svg')
plt.show()
```

```python
# convert radians to degrees

n = 10
rad = np.logspace(np.log10(1),np.log10(2*np.pi),n) 

rad = np.linspace(0,2*np.pi,n) #I like this way better

print(rad)
print(np.rad2deg(rad))
```

```python
# famous equality in trigonometry

ang = np.logspace(np.log10(1),np.log10(2*np.pi),10)
np.cos(ang)**2 + np.sin(ang)**2
```

```python
# create euler's number
p = np.pi/4
m = .5

eulr = m*np.exp(1j*p)

# now extract magnitude and phase
mag = np.abs(eulr)
ang = np.angle(eulr)

print(mag, ang)

# then plot
plt.polar([0,ang], [0,mag],'b',linewidth=3)
plt.show()
```
