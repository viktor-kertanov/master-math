```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler_from_sin_cos():
    cosine_part = eval(input('cosine part: '))  
    sine_part = eval(input('sine part: '))
    magnitude = np.abs(cosine_part + 1j*sine_part)
    angle = np.angle(cosine_part + 1j*sine_part)

    plt.polar([0, angle], [0, magnitude], 'b', linewidth=2)
    plt.thetagrids([45, 130, 215, 250, 0])
    plt.title('$me^{i\phi}, m=%g, k=%g$' % (magnitude, angle))
    plt.show()

euler_from_sin_cos()


 
```
