```python
import numpy as np
import matplotlib.pyplot as plt
```

There's some point and there's a whole theory behind that: after a certain point of resolution (how many "dots" we connect on the graph) - nothing is changing. The code only take a little bit longer to run. So how can we find the sweet point?

```python
x = np.linspace(0, 6*np.pi, 100)

plt.plot(x, np.sin(x), 'r', label='sin($\\theta$')
plt.plot(x, np.cos(x), 'b', label='cos($\\theta$)')

plt.xlabel('rad')
plt.ylabel('function value')
plt.legend()
plt.show()
```

```python
th = np.linspace(0, 4*np.pi, 300)

plt.plot(th, np.tan(th), 'k')

#zoom in
plt.ylim([-5,5])
plt.show()
```

```python
angle = np.random.rand()*2*np.pi

tan = np.tan(angle)
tan_2 = np.sin(angle) / np.cos(angle)

print(f"tan: {tan}. tan_2: {tan_2}")
```
