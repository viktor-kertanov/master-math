```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nums = np.random.rand(1000)

plt.plot(nums, 'o', linewidth=0.5)
plt.show()

plt.hist(nums)
plt.show()
```

If we need to shift our uniform distribution and center it around another number we can just add the center to each point in matrix:

```python
nums = np.random.rand(1000) + 3

plt.plot(nums, 'gs')
plt.show()

plt.hist(nums, color='green')
plt.show()
```

To generate uniformly distributed random numbers in a certain range [a, b] we need to:
1) generate a vector of a certain size with number in [0, 1]
2) multiply each number in the vector by (b-a)
3) add a

```python
nums = np.random.rand(20)
minval = 2
maxval = 17

nums_final = nums*(maxval-minval) + minval

plt.plot(nums_final, 'rs')
plt.show()

plt.hist(nums_final, color='red')
plt.show()
```

```python
normal_distr_nums = np.random.randn(100000)

plt.plot(normal_distr_nums, 'bo', alpha=0.01) # with alpha transparency this looks just amazing
plt.show()

plt.hist(normal_distr_nums, color='blue')
plt.show()
```
