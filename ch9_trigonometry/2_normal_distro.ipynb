```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mean = 15, sigma = 4.3
mean = 15
standard_deviation = 4.3

nums = np.random.normal(mean, standard_deviation, 10000)

plt.plot(nums, 'go', alpha=0.1)
plt.show()

plt.hist(nums, color='tab:green')
plt.show()
```

```python
# Another solution to the same task
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(10000)*4.3 + desired_mean

nums = nums/np.std(nums) # to sed the mean to match exactly 1 ("standard normal")
nums = nums * desired_std

nums = nums - np.mean(nums) # cool way to force the mean to exactly to ZERO
nums = nums + desired_mean # now setting from ZERO to DESIRED MEAN

plt.subplot(121)
plt.plot(nums, 's', alpha=0.1)

plt.subplot(122)
plt.hist(nums)

plt.show()

print('This distribution has a mean of %s and standrad deviation of %s' % (np.mean(nums), np.std(nums)))
```
