```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
angles = np.linspace(0, 2*np.pi, 1000)

x = np.cos(angles)
y = np.sin(angles)

n=1000
random_angles = np.random.rand(n)*2*np.pi

# plt.plot(x, y)
for angle in random_angles:
    plt.plot([0, np.cos(angle)], [0, np.sin(angle)])
plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
#Proof check
n=100000
random_angles = np.random.rand(n)*2*np.pi

plt.plot(random_angles, 'ro', alpha=0.009)
plt.plot([0, n], [np.mean(random_angles), np.mean(random_angles)], 'g--', linewidth=1)
plt.show()

print(f'Our mean angles is: {np.mean(random_angles)}. STD: {np.std(random_angles)}')
```

```python
# The solution from the video

angles = np.random.rand(1000)*2*np.pi

for angle in angles:
    plt.plot([0, np.cos(angle)], [0, np.sin(angle)], 'k', linewidth=0.5)

plt.axis('square')
plt.axis([-1, 1, -1, 1])
plt.axis('off')
plt.show()
```
