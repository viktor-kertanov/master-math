# Euler's formula, or "Euler's Identity"

Euler's formula leads to what know as "the most elegant equation" in human civilization.
You can represent any point on the complex plain using this equation:

m $\times \exp^{ik}$ = m $\times$ (cos(k) + i sin(k))

```python
k = np.pi / 6
m = 2.3

eulers_formula_l = m * np.exp(1j*k)
eulers_formula_r = m * (np.cos(k) + 1j * np.sin(k))

print(f"Left side: {eulers_formula_l}. Right side: {eulers_formula_r}")
```

```python
magnitude = np.abs(eulers_formula_l)

angle = np.angle(eulers_formula_l)

print(magnitude, angle)

plt.polar([0, angle], [0, magnitude], 'r')
plt.polar(k, m, 'bo')
plt.show()
```
