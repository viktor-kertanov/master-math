```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Math, display
import sympy as sym
```

```python
def angle_convert_plot():
    user_input, measure = None, None
    while not user_input:
        try:
            user_input = eval(input('What angle to convert?').strip().replace(" ", "")) #wow, eval is better than float
        except ValueError:
            continue

    while measure != 'd' and measure != 'r':
        measure_input = input('Which unit did you input (radian or degree, "r"/"d")?')
        measure = measure_input.lower()[0]
    
    if measure == 'd':
        input_angle_deg = user_input
        input_angle_deg_optimised = user_input % 360
        input_angle_rad_optimised = np.deg2rad(input_angle_deg_optimised)
        measure_str = 'DEG'
    if measure == 'r':
        input_angle_rad = user_input
        input_angle_rad_optimised = user_input % (2*np.pi)
        input_angle_deg_optimised = np.rad2deg(input_angle_rad) % 360
        measure_str = 'RAD'
    
    #full circle
    dots = np.linspace(0, 2*np.pi, 1000)
    circle_x = np.cos(dots)
    circle_y = np.sin(dots)
    plt.plot(circle_x, circle_y, 'b', linewidth=1)

    # input angle radius
    plt.plot([0, np.cos(0)], [0, np.sin(0)], 'g', linewidth=2)
    plt.plot([0, np.cos(input_angle_rad_optimised)], [0, np.sin(input_angle_rad_optimised)], 'g', linewidth=2)

    angle_circle_measure = np.linspace(0, input_angle_rad_optimised, 100)
    radii = [0.1, 0.08, 1]
    for r in radii:
        x = np.cos(angle_circle_measure)*r
        y = np.sin(angle_circle_measure)*r
        plt.plot(x, y, 'g', linewidth=2)

    

    plt.gca().set_aspect('equal')
    plt.grid()
    plt.title('Your input was %g %s.\nThis is equal to angle of $%s^\circ$DEG or $%g$ RAD' % (user_input, measure_str, input_angle_deg_optimised, input_angle_rad_optimised))
    plt.axis([-1.1, 1.1, -1.1, 1.1])
    
    plt.show()

angle_convert_plot()
```
