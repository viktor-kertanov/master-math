```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mexp^(ik), m, k randomly generated

pink = [1, .2, .7]
purple = [.7, .2, 1]
green = [0, 1, 0]

def get_color():
    number = np.random.rand()
    if number <= 0.4:
        return pink
    elif number > 0.4 and number <= 0.8:
        return purple
    else:
        return green

num_lines = 200
m = np.random.rand(num_lines)*5 + 2
k = np.random.rand(num_lines)*2*np.pi

for idx in range(num_lines):
    plt.polar([0, k[idx]], [0, m[idx]], color=get_color(), linewidth=1.5)

plt.axis('off')
plt.show()
```
