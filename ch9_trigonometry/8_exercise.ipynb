```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import Math, display
```

```python
thetas = np.linspace(-6*np.pi, 6*np.pi, 1000)

fxy = np.sin(thetas + np.cos(thetas))
fxy_2 = np.cos(thetas + np.sin(thetas))

plt.plot(thetas, fxy, 'r', linewidth=2, label='$sin(\\theta+cos(\\theta))$')
plt.plot(thetas, fxy_2, 'b', linewidth=2, label='$cos(\\theta+sin(\\theta))$')

plt.legend()
plt.show()
```

```python
# input_theta = sym.symbols('theta')
theta = sym.pi/4
input_theta = float(theta)


# unit circle
circle_points = np.linspace(0, 2*np.pi, 1000)
plt.plot(np.cos(circle_points), np.sin(circle_points), 'k', alpha=0.3)

xlim = plt.gca().get_xlim()
ylim = plt.gca().get_ylim()

plt.plot([xlim[0], xlim[1]], [0, 0], 'k--', alpha=0.3, linewidth=1)
plt.plot([0, 0], [ylim[0], ylim[1]],  'k--', alpha=0.3, linewidth=1)

# drawing the main arch
arch_dots = np.linspace(0, input_theta, 100)
plt.plot(np.cos(arch_dots), np.sin(arch_dots), 'k', linewidth=2)
plt.plot(np.cos(input_theta), np.sin(input_theta), 'ko')

# aux lines
plt.plot([np.cos(input_theta), 0],[np.sin(input_theta), np.sin(input_theta)], 'r--', linewidth=0.5)
plt.plot([0, 0],[0, np.sin(input_theta)], 'r', linewidth=2, label='sin part')

plt.plot([np.cos(input_theta), np.cos(input_theta)],[np.sin(input_theta), 0], 'g--', linewidth=0.5)
plt.plot([0, np.cos(input_theta)],[0, 0], 'g', linewidth=2, label='cos part')

plt.plot([0, np.cos(input_theta)], [0, np.sin(input_theta)], 'k:', linewidth=0.5, alpha=0.9)

plt.gca().set_aspect('equal')
plt.axis([xlim[0], xlim[1], ylim[0], ylim[1]])

plt.title('$\\theta = %s$' % sym.latex(theta))
plt.legend(loc='lower left')
plt.show()
```
