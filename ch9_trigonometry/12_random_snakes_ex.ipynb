```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r1 = np.random.rand()
r2 = np.random.rand()

points = np.linspace(-24*np.pi, 24*np.pi, 10000)

x = np.cos(r1*points)
y = np.sin(r2*points)

plt.subplot(121)
plt.plot(x, y, 'k', linewidth=1)
plt.plot([np.cos(r1*points[0]), np.cos(r1*points[-1])], [np.sin(r2*points[0]), np.sin(r2*points[-1])], 'r', linewidth=0.5)
plt.title(f'r$_1$={r1:.2f}, r$_2$={r2:.2f}')
plt.axis('off')

plt.subplot(122)
plt.plot(points, x, points, y)

plt.show()
```
