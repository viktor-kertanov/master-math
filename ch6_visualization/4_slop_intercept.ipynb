```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The slope-intercept form is y = mx+b, standard mathmatics
# m coeff is a "slope"; b coeff - is intercept parameter, i.e. where the line crosses y-axis when x value is 0.
```

```python
x = [-5, 5]
m = 2
b = 1

y = [p*m+b for p in x] #list comprehension
y = m*np.array(x)+b #numpy array

#plotting
plt.plot(x,y, label=f'y={m}x+{b}')

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

#axis and coordinates
axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_xlim(), 'k--')

plt.legend()

plt.show()
```

```python
#exercise
x = [-5, 5]
m1 = 0.7
b1 = -2

m2 = -1.25
b2 = 0.75
```

```python
y1 = m1*np.array(x) + b1
y2 = m2*np.array(x) + b2
```

```python
plt.plot(x, y1, label=f'y={m1}x+{b1}')
plt.plot(x, y2, label=f'y={m2}x+{b2}')

plt.axis('square')
plt.axis([x[0],x[1],-5,5])
plt.grid()

axis = plt.gca()
plt.plot(axis.get_xlim(),[0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
