```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np 
from IPython.display import display, Math
import sympy as sym
```

```python
y = np.array([
    [1,1],
    [2,3],
    [3,1 ]
])

display(Math(sym.latex(sym.sympify(y))))
```

```python
y1 = np.array([
    [2,2],
    [2.5, 4],
    [3.5, 1]
])
display(Math(sym.latex(sym.sympify(y1))))
```

```python
p = Polygon(y, facecolor='m', alpha=0.3)
p1 = Polygon(y1, facecolor='None', alpha=1, edgecolor='k', linewidth=0.5)

fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```

```python
#Exercise: Hobbit house, f(x) = -x^2
```

```python
n = 100
x = np.linspace(-2,2,n)
y = [(el, -el**2) for el in x]
Y = np.array(y)

#square
sq_array = np.array([
    [-0.5,-4],
    [-0.5,-2.5],
    [0.5,-2.5],
    [0.5,-4]
])

# display(Math(sym.latex(sym.sympify(Y))))
```

```python
parabola = Polygon(Y, facecolor='lightgreen', alpha=0.5, edgecolor='k', linewidth=2)
square = Polygon(sq_array, facecolor='k', alpha=1)

fig, ax = plt.subplots()
ax.add_patch(parabola)
ax.add_patch(square)

ax.set_xlim([-2.2,2.2])
ax.set_ylim([-4.2,0.2])

plt.show()
```
