```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# In squircle let's start with x^4 + y^4 = a
a = 1

x = np.linspace(-4,4, 1000)
y = np.linspace(-4, 4, 1000)

for a in np.logspace(np.log10(0.01), np.log(100), 50):
    X, Y = np.meshgrid(x, y)
    Fxy = X**4 + Y**4 - a

    plt.contour(X, Y, Fxy, 0, linewidths=1, colors=(np.random.rand(), np.random.rand(), np.random.rand()))
    plt.axis('square')
plt.axis('off')
plt.savefig('squircles.svg')
plt.show()
```

```python
# In squircle let's start with x^4 + y^4 = a
a = 1

for a in np.arange(1, 6, 5):
    x = np.linspace(-a,a, 2001)
    y = (a - x**4)**(0.25)
    plt.plot(x, y, 'k')
    plt.plot(x, -y, 'k')

plt.axis('off')
plt.axis('square')
plt.savefig('squircles.svg')
plt.show()
```
