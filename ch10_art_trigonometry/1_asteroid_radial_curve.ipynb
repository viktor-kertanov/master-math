```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig_num = 24
col_in_one_row = 4
for n in range(fig_num):

    pi_scale = 12
    steps = 10009
    t = np.linspace(-pi_scale*np.pi, pi_scale*np.pi, steps)
    a = t
    #subplots

    plt.subplot(fig_num//col_in_one_row+1, col_in_one_row, n+1)
    plt.plot(a**n*np.cos(t)**3, a**n*np.sin(t)**3, 'k', linewidth=0.5)
    # plt.gca().set_aspect('equal')
    plt.axis('square')
    plt.axis('off')
    # plt.title(f'n={n}', fontsize=6)
      
plt.savefig('8_asteroid_radial_curve.svg')
plt.show()
```

```python
fig_num = 9
col_in_one_row = 3

# Another way to solve the subplot challenge
fig, ax = plt.subplots(fig_num//col_in_one_row, col_in_one_row, figsize=(15, 10))

for n in range(fig_num):
    pi_scale = 6
    steps = 10000
    t = np.linspace(-pi_scale*np.pi, pi_scale*np.pi, steps)
    a = t
    #indexes
    subplot_row = n // col_in_one_row
    subplot_col = n % col_in_one_row
    #subplots
    ax[subplot_row, subplot_col].plot(a**n*np.cos(t)**3, a**n*np.sin(t)**3, 'k', linewidth=0.5)
    ax[subplot_row, subplot_col].axis('square')
    ax[subplot_row, subplot_col].axis('off')
    # plt.title(f'n={n}', fontsize=6)
      
plt.savefig('8_asteroid_radial_curve.svg')
plt.axis('off')
plt.show()
```
