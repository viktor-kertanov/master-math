# Logistic Maps

The logistic map formula:

$\newline x_{n+1} = rx_n(1-x_n) \newline$

*Conditions*:
$\newline x_1 = .5 \newline
n \in \mathbb{N}: n<=500 \newline
r \in [1, 4] \text{ in 1000 steps} \newline$

First we iterate over n for a certain r (e.g. r=1), then we create x vectors for each r.

After we've created the vectors, for each value **r** we plot unique values in the last **10% of the vector x**.<br>
**R** goes on the *x-axis*, **x** goes on the *y-axis*.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Now for one vector
x_size = 500
size_10 = int(np.floor(x_size*0.1))
x_init = 0.5
r=1
# Vector X
x = np.zeros(x_size)
x[0] = x_init
for idx, el in enumerate(x[1:], start=1):
    x[idx] = r*x[idx-1]*(1-x[idx-1])

#Last 10% & unique of the vector x
unique_last_10_x = np.unique(x[-size_10:])

#plot
plt.plot(x, 's-', linewidth=0.5, alpha=0.5)

# plt.axis('square')
# plt.axis([3, 4, 0, 1])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.show()
```

```python
x_size = 1000
size_10 = int(np.floor(x_size*0.1))
x_init = 0.5

for r in np.linspace(1, 4, 1000):
    # Vector X
    x = np.zeros(x_size)
    x[0] = x_init
    for idx, el in enumerate(x[1:], start=1):
        x[idx] = r*x[idx-1]*(1-x[idx-1])

    #Last 10% & unique of the vector x
    unique_last_10_x = np.unique(x[-size_10:])

    #plot
    plt.plot(np.full(unique_last_10_x.shape, r), unique_last_10_x, 'k.', markersize=0.5, linewidth=0.5, alpha=0.3)

# plt.axis('square')
# plt.axis([3, 4, 0, 1])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.show()
```

```python
x_size = 1000
size_10 = int(np.floor(x_size*0.1))
x_init = 0.5

for r in np.linspace(1, 4, 1000):
    # Vector X
    x = np.zeros(x_size)
    x[0] = x_init
    for idx, el in enumerate(x[1:], start=1):
        x[idx] = r*x[idx-1]*(1-x[idx-1]) + .2*np.sin(2*np.pi*x[idx-1])**2

    #Last 10% & unique of the vector x
    unique_last_10_x = np.unique(x[-size_10:])

    #plot
    plt.plot(np.full(unique_last_10_x.shape, r), unique_last_10_x, '.', markersize=2, alpha=0.5)

# plt.axis('square')
# plt.axis([3, 4, 0, 1])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.axis('off')
plt.savefig('the_logistic_map.svg')
plt.show()
```
