```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-4*np.pi, 4*np.pi, 3000)

f1 = np.cos(t)
# f2 = np.exp(-t)

plt.plot(t, f1)
# plt.plot(t, f2)

# plt.axis([-, 2, -2, 2])
plt.show()
```

```python
pi_scale = 10
t = np.linspace(0, pi_scale*np.pi, 2000)

k = -3
x = np.cos(t)*np.exp(t/k)
y = np.sin(t)*np.exp(t/k)

plt.plot(x, y, linewidth=1, color='r')
plt.gca().set_aspect('equal')
# plt.axis('off')
plt.axis('square')
plt.savefig('log_spirals.svg')
plt.show()

```

```python
pi_scale = 12
t = np.linspace(0, pi_scale*np.pi, 3000)

spirals_num = 50

for i_idx, i in enumerate(np.linspace(2, 6, spirals_num)):
    k = -i
    x = np.cos(t)*np.exp(t/k)
    y = np.sin(t)*np.exp(t/k)

    plt.plot(x, y, linewidth=4, color=(1-i_idx/spirals_num, 0, 1-i_idx/spirals_num))
plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('log_spirals.svg')
plt.show()
```

```python
pi_scale = 5
t = np.linspace(0, pi_scale*np.pi, 1000)

spirals_num = 40

for i_idx, i in enumerate(np.linspace(0.01, 6, spirals_num)):
    k = -i
    x = np.cos(t)*np.exp(t/k) + .018*np.random.rand()*np.cos(2*np.pi*t)**2
    y = np.sin(t)*np.exp(t/k) + .018*np.random.rand()*np.sin(2*np.pi*t)**2

    plt.plot(x, y, linewidth=0.5, color='k')
plt.gca().set_aspect('equal')
plt.axis('off')
plt.gcf().set_size_inches(18.5, 10.5)
plt.savefig('log_spirals_print_print_random_50.svg')
plt.show()
```

```python
spirals_num = 300

for k_idx, k in enumerate(np.linspace(0.01, 8, spirals_num)):
    pi_scale = (k_idx / spirals_num)*1.3
    t = np.linspace(0, pi_scale*np.pi, 2000)
    x = np.cos(t)*np.exp(t/-k) + .08*np.random.rand()*np.cos(2*np.pi*t)**2
    y = np.sin(t)*np.exp(t/-k) + .08*np.random.rand()*np.sin(2*np.pi*t)**2

    plt.plot(x, y, linewidth=0.5, color='k')
plt.gca().set_aspect('equal')
plt.axis('off')
plt.gcf().set_size_inches(18.5, 10.5)
plt.savefig('log_spirals_print_print.svg')
plt.show()
```
