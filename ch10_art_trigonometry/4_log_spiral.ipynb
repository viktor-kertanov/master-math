```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-4*np.pi, 4*np.pi, 3000)

f1 = np.cos(t)
# f2 = np.exp(-t)

plt.plot(t, f1)
# plt.plot(t, f2)

# plt.axis([-, 2, -2, 2])
plt.show()
```

```python
pi_scale = 6
t = np.linspace(0, pi_scale*np.pi, 3000)

spirals_num = 40

for i in np.logspace(np.log10(0.1),np.log10(100), spirals_num):
    k = -i
    x = np.cos(t)*np.exp(t/k)
    y = np.sin(t)*np.exp(t/k)
    # x = np.cos(t)*np.exp(t/k)*np.random.rand()+np.random.rand()
    # y = np.sin(t)*np.exp(t/k)*np.random.rand()+np.random.rand()

    plt.plot(x, y, linewidth=5, color=(i/spirals_num, 0, i/spirals_num))
    plt.gca().set_aspect('equal')
    plt.axis('off')
plt.savefig('log_spirals.svg')
plt.show()

```
