```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
steps = 1000
x_min = -3
x_max = 3
x = np.linspace(x_min, x_max, steps)
y_1 = x
y_2 = x**2
y_3 = x**3

y_h1, x_h1 = np.histogram(y_1, bins=steps)
y_h2, x_h2 = np.histogram(y_2, bins=steps)
y_h3, x_h3 = np.histogram(y_3, bins=steps)

fig, ax = plt.subplots(1, 3)

ax[0].bar(x_h1[:-1], y_h1/np.sum(y_h1))
ax[0].set_xticks(np.arange(x_min, x_max, 3))
ax[0].set_title('y=x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('probability')

ax[1].bar(x_h2[:-1], y_h2/np.sum(y_h2))
# ax[1].set_xticks(np.arange(x_min, x_max, 3))
ax[1].set_title('$y=x^2$')
ax[1].set_xlabel('x')
ax[1].set_ylabel('probability')

ax[2].bar(x_h3[:-1], y_h3/np.sum(y_h3))
# ax[1].set_xticks(np.arange(x_min, x_max, 3))
ax[2].set_title('$y=x^3$')
ax[2].set_xlabel('x')
ax[2].set_ylabel('probability')

plt.subplots_adjust(wspace=0.3)
plt.gcf().set_size_inches(13, 8)
plt.show()
```

```python
steps = 1000
x_min = -3
x_max = 3

x = np.linspace(x_min, x_max, steps)

functions = [x, x**2, x**3]
colors = ['r', 'g', 'b']
func_names = ['y=x', '$y=x^2$', '$y=x^3$']
fig, ax = plt.subplots(2,  len(functions))

for fx_idx, fx in enumerate(functions):
    y_hist, x_hist = np.histogram(fx, bins=50)
    
    ax[0][fx_idx].bar(x_hist[:-1], y_hist/np.sum(y_hist), color=colors[fx_idx]) # bar/plot
    ax[0][fx_idx].set_ylabel('probability')
    ax[1][fx_idx].plot(x, fx, color=colors[fx_idx])
    ax[1][fx_idx].set_title(func_names[fx_idx])
    ax[1][fx_idx].set_xlabel('x')
    ax[1][fx_idx].set_ylabel('fx')

plt.subplots_adjust(wspace=0.3)
plt.gcf().set_size_inches(13, 8)
plt.show()
```
