```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num_trials = 50
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [2/7, 1/7, 1/7, 1/7, 1/7, 1/7]

random_gen = np.random.choice(outcomes, num_trials, p=probabilities)

fx, x = np.histogram(random_gen, bins=np.arange(1, 8))

# print(fx, x)

fx_p = fx/np.sum(fx)
plt.bar(x[:-1], fx_p, color='k', label='Empirical (proportion)')
plt.plot(outcomes, probabilities, 'ms-', markersize=10, label='Theoretical (probability)')

for idx in range(6):
    plt.text(x[idx], fx_p[idx]-0.02, '{:.0%}'.format(fx_p[idx]), color='w', ha='center')

plt.ylabel('Probability/Proportion')
plt.xlabel('Outcomes')
plt.legend()
plt.gcf().set_size_inches(10,7)
plt.show()


```
