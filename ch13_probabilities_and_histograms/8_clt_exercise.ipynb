```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
```

```python
ns = np.arange(50, 10000, 100)
num_bins = 100

all_hist = np.zeros((len(ns), num_bins))
print(all_hist.shape)
for i in range(len(ns)):
    sig_1 = np.sin(np.linspace(0, 6*np.pi, ns[i]))
    sig_2 = np.random.rand(ns[i])*2 - 1
      
    all_hist[i, :], x= np.histogram(sig_1 + sig_2, bins=num_bins, density=True)  

plt.pcolor(x[:-1], ns, all_hist, vmax=.6)
plt.xlabel('Data value')
plt.ylabel('Points in the sample')
plt.colorbar()
plt.show()
```
