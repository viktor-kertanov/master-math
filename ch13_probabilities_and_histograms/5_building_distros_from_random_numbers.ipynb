### __Building distributions from random numbers__

- Hot to construct a histogram of a random process
- Several different distribution shapes
- How to normalize any distribution into a Gaussian distribution

To this point we were working with discrete distributions, but not everything is discrete (like a length of a petal or a grass etc.)

Coin or dice: examples of __discrete processes__.
There are also __random processes__, values can be any, not only from our sample.

In random processes we use ranges (e.g. 1-2; 2-3 bins). 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
shift = 5
stretch = 2
num = 100000

points = np.random.randn(num)*stretch + shift 
print(f'Mean of random data points: {np.mean(points)}. Standard deviation: {np.std(points)}')

fig, ax = plt.subplots(1, 2, figsize=(6, 4))

ax[0].plot(points, 's', alpha=.4)
ax[1].hist(points, bins=100) #we increase the number of bins to increase resolution

plt.subplots_adjust(wspace=.3)
plt.gcf().set_size_inches(13, 7)
plt.show()
```

Below is the __uniform distribution__. 

```python
shift = 5
stretch = 2
num = 100000

points = np.random.rand(num)*stretch + shift - .5 * stretch
print(f'Mean of random data points: {np.mean(points)}. Range: [{np.max(points) - np.min(points)}]')

fig, ax = plt.subplots(1, 2, figsize=(6, 4))

ax[0].plot(points, 'sg', alpha=.01)
ax[1].hist(points, bins=[4, 4.5, 5, 5.5, 6], color='g', edgecolor='k') #we increase the number of bins to increase resolution

plt.subplots_adjust(wspace=.3)
plt.gcf().set_size_inches(13, 7)
plt.show()
```

Now to the __Poisson distribution__.
We see that outputs are all __integers__ and distributions looks a bit like a Gaussian but it is shifted towards left.

__Lambda__ parameter in Poissson distribution encodes __both__ the mean and the variance!

So we get the bunch of integers in the output which has the __mean of lambda__ and a __variance of lambda__ which is clearly seen in the results below.

```python
poisson_lambda = 3.4
stretch = 2
num = 10000

points = np.random.poisson(lam=poisson_lambda, size=num)
print(f'Mean is: {"{:.1f}".format(np.mean(points))}. Variance is: {"{:.1f}".format(np.var(points))}')

fig, ax = plt.subplots(1, 2, figsize=(6, 4))

ax[0].plot(points, 'sr', alpha=.01)
ax[1].hist(points, bins=np.arange(np.max(points)+1), color='r', edgecolor='k') 

plt.subplots_adjust(wspace=.2)
plt.gcf().set_size_inches(13, 7)
plt.show() 
```
