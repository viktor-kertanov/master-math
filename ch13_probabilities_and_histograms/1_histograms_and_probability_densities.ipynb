### __Probabilities and histograms__

It is important to know that __proportion__ and __probabilities__ are different things and have different interpretations.
Proportion shows how many % in our data set A or B or C occurred. Probability shows what is the chance to pick A or B or C if we pick totally randomly. A may have 25% chance to get picked. B - 50%; C - 25%.
<br>
Proportions operates on factual appearances (6 out of 10 coin flips landed on heads; 60% proportion of heads). Probability though of heads to appear on any given flip is __not 60%, but 50%__.  

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data = [1, 2, 2, 2, 2, 2, 3, 4, 5, 6, 6, 6, 6, 6, 7, 7, 7, 8, 6, 3, 3, 3, 2, 1, 7, 8, 9, 8, 9, 1, 1, 1]

plt.hist(data)
plt.xlabel('Values')
plt.ylabel('Number of Appearances (count)')

# plt.grid()
plt.axis([1, 10, 0, 10])
plt.show()

histogram_data = np.histogram(data)
print(histogram_data)
```

What we see above, specifically __histogram data__ output, looks seemingly strange, by default hist/histogram function tries to pack the data into 10 "bins".
We see the empty space in the middle of the histogram which means that __in the range between [4.2, 5) (right boundary exclusive) there are no data points__.

```python
data = [1, 2, 2, 2, 2, 2, 3, 4, 4, 4, 5, 6, 6, 6, 6, 6, 7, 7, 7, 8, 6, 3, 3, 3, 2, 1, 7, 8, 9, 8, 9, 1, 1, 1, 1, 1, 1 ]

y, x = np.histogram(data, bins=np.arange(1, 11))
print(f'Y-values (counter): {y}, X-values: {x}')

fig, ax = plt.subplots(1, 3)

# Drawing proportion
ax[0].bar(x[:-1], y)
ax[0].set_xlabel('Values')
ax[0].set_ylabel('Number of Appearances (count)')
ax[0].set_xticks(np.arange(1, 10, 1))

# Calculating probability
y_probability = y/np.sum(y)
print(f'Total probability sum of y-values: {np.sum(y_probability)}')

# Drawing probability density plot
ax[1].bar(x[:-1], y_probability)
ax[1].set_xlabel('Values')
ax[1].set_ylabel('Probability density')
ax[1].set_xticks(np.arange(1, 10, 1))

# Normalizing the probability -- the maximum becomes value 1
y_probability_norm = y/np.max(y)
ax[2].bar(x[:-1], y_probability_norm)
ax[2].set_xlabel('Values')
ax[2].set_ylabel('Proportion/probability (normalized)')
ax[2].set_xticks(np.arange(1, 10, 1))


plt.subplots_adjust(wspace=0.5)
plt.gcf().set_size_inches(10, 7)
plt.show()

```
