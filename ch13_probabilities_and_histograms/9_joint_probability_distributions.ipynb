### __Joint Probability Distributions__

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
n = 1000

x = np.random.randn(n)
y = np.random.randn(n)**2

plt.hist2d(x, y, bins=30, vmax=10)
plt.xlabel('x')
plt.ylabel('$x^2$') 

plt.show()
```

```python
sns.set()
sns.jointplot(x=x, y=y, kind='scatter', color=[.8, .3, .9]).plot_joint(sns.kdeplot)
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.show()
```

```python
df = pd.read_csv('probability_data.csv', header=None, names=['x', 'y'])
x = df['x']
y = df['y']
sns.jointplot(x=x, y=y, kind='scatter', color=[.3, .8, .5]).plot_joint(sns.kdeplot)
plt.xlabel('x')
plt.ylabel('$y$')
plt.show()
```
