### __Central Limit Theorem__

Very important theorem that has a lot of practical applications.

__CLT__:

__Random samples from independent variables will tend towards a normal distribution (Gaussian), even if the variables are non-normally distributed__

<br>

__Example:__
if we have a sine wave (non-normally distributed) & uniform random distribution (non-normally distributed) and combine them together, the result will tend towards normal distribtuion. And that's remarkable.

There are some conditions:
1. We need quite a lot of data points
2. The distributions should be in the same scale of values
3. We may need to apply some proper normalization techniques for this to work

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 30000

signal_1 = np.sin(np.linspace(0, 6*np.pi, n))
signal_2 = np.random.rand(n)*2 -1

combined = signal_1 + signal_2
#plotting the individual signals and its distributions

fig, ax = plt.subplots(2, 3, figsize=(10,4))

# Sine wave
ax[0, 0].plot(signal_1) 
ax[0, 0].set_title("Sine wave")
ax[1, 0].hist(signal_1, bins=100)

# Unifrom distribution
ax[0, 1].plot(signal_2, 's', alpha=.005)
ax[0, 1].set_title("Uniform distribution")
ax[1, 1].hist(signal_2, bins=100)

# Combined signal
ax[0, 2].plot(combined, 's', alpha=.005)
ax[0, 2].set_title("Combined signal")
ax[1, 2].hist(combined, bins=100)

for i in np.matrix.flatten(ax):
    i.set_xticks([])
    i.set_yticks([])

plt.subplots_adjust(wspace=0.3)
plt.show()
```

Combining 2 and more signals not necessarily creates a distributions that tends towards a normal distribution. Below is the example when we change the scale of signal 2, but the result is very close to a sine wave. 
The data should be in the same scale!

```python
n = 30000

signal_1 = np.sin(np.linspace(0, 6*np.pi, n))
signal_2 = (np.random.rand(n)*2 -1) / 10

combined = signal_1 + signal_2
#plotting the individual signals and its distributions

fig, ax = plt.subplots(2, 3, figsize=(10,4))

# Sine wave
ax[0, 0].plot(signal_1) 
ax[0, 0].set_title("Sine wave")
ax[1, 0].hist(signal_1, bins=100)

# Unifrom distribution
ax[0, 1].plot(signal_2, 's', alpha=.005)
ax[0, 1].set_title("Uniform distribution")
ax[1, 1].hist(signal_2, bins=100)

# Combined signal
ax[0, 2].plot(combined, 's', alpha=.005)
ax[0, 2].set_title("Combined signal")
ax[1, 2].hist(combined, bins=100)

for i in np.matrix.flatten(ax):
    i.set_xticks([])
    i.set_yticks([])

plt.subplots_adjust(wspace=0.3)
plt.show()
```
