# Polygons
#### How to create a proper polygon of with any number of sides of a certain length given the centers coordinates

```python
center_x = 0
center_y = 0

radius = 2
scale = 1.1
n=8

angles = np.linspace(0, 360, n+1)
x = radius*np.cos(np.deg2rad(angles))
y = radius*np.sin(np.deg2rad(angles))

# for angle in angles:
#     plt.plot([center_x, radius*np.cos(np.deg2rad(angle))], [center_y, radius*np.sin(np.deg2rad(angle))], 'k', linewidth=1)

plt.plot(x,y, 'k', linewidth=1)
plt.axis('square')
plt.plot([-radius*scale, radius*scale], [0, 0], 'k--', alpha=0.5, linewidth=1)
plt.plot([0,0], [-radius*scale, radius*scale], 'k--', alpha=0.5, linewidth=1)
plt.grid()
plt.show()
```

```python
# drawing circle
circle_angles = np.linspace(0, 2*np.pi, 100)
radius = 10
circle_x = radius*np.cos(circle_angles)
circle_y = radius*np.sin(circle_angles)

plt.plot(circle_x, circle_y)

#m points on circle
m=24
angles = np.linspace(0, 360, m+1)
m_x = radius*np.cos(np.deg2rad(angles))
m_y = radius*np.sin(np.deg2rad(angles))
plt.plot(m_x, m_y, 'ro', markersize=2)

# m polygons around each m point
for p_idx in range(m):
    center_x = m_x[p_idx]
    center_y = m_y[p_idx]
    p_radius = 2
    n=6

    angles = np.linspace(0, 360, n+1)
    x = center_x + p_radius*np.cos(np.deg2rad(angles))
    y = center_y + p_radius*np.sin(np.deg2rad(angles))

    plt.plot(x,y, 'k', linewidth=1)


plt.gca().set_aspect('equal')
plt.show()
```

Good idea to understand how to rotate polygons around the center of a polygon

```python
# polygon
center_x = 0
center_y = 0

radius = 2
scale = 1.1
n=3

angles = np.linspace(0, 360, n+1)
x = radius*np.cos(np.deg2rad(angles))
y = radius*np.sin(np.deg2rad(angles))


# rotation
t = 12 # theta points
theta_points = np.linspace(0, 360, t+1)[:-1]
print(theta_points)
for theta_deg in theta_points:
    theta = np.deg2rad(theta_deg)
    sin_theta, cos_theta = np.sin(theta), np.cos(theta)
    
    x_r = cos_theta*x - sin_theta*y
    y_r = sin_theta*x + cos_theta*y

    plt.plot(x_r, y_r, color='k', label=f'$\\theta$={theta_deg}')

# plt.plot(x,y, 'k', linewidth=1)

plt.axis('square')

# grid and coordinates plane
plt.plot([-radius*scale, radius*scale], [0, 0], 'k--', alpha=0.5, linewidth=1)
plt.plot([0,0], [-radius*scale, radius*scale], 'k--', alpha=0.5, linewidth=1)
plt.grid()


plt.show()
```
