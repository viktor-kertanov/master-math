```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from datetime import datetime
```

Now the idea is to rotate a little bit each triangle around its center.
Let's start small with small number of points

```python
n_points_circle=180
circle_scale = 40
theta = np.linspace(0, 2*np.pi, n_points_circle)
circle_x = circle_scale*np.cos(theta)
circle_y = circle_scale*np.sin(theta)

triangle_median = 4

# plt.plot(circle_x, circle_y, 'bo')

for point_idx in range(n_points_circle):
    triangle_center_x = circle_x[point_idx]
    triangle_center_y = circle_y[point_idx]

    alpha_angle = np.pi/2

    A = (
        triangle_center_x * alpha_angle,
        (triangle_center_y + triangle_median) * alpha_angle
    )
    B = (
        (triangle_center_x + (np.sqrt(3)*triangle_median)/2) * alpha_angle,
        (triangle_center_y - triangle_median/2) * alpha_angle
    )
    C = (
        (triangle_center_x - (np.sqrt(3)*triangle_median)/2) * alpha_angle,
        (triangle_center_y - triangle_median/2) * alpha_angle
    )

    plt.plot([A[0],B[0]], [A[1], B[1]], 'k', linewidth=0.5)
    plt.plot([A[0],C[0]], [A[1], C[1]], 'k', linewidth=0.5)
    plt.plot([B[0],C[0]], [B[1], C[1]], 'k', linewidth=0.5)
    # plt.plot(triangle_center_x, triangle_center_y, 'ro')

    A = (
        triangle_center_x ,
        triangle_center_y + triangle_median
    )
    B = (
        triangle_center_x + (np.sqrt(3)*triangle_median)/2,
        triangle_center_y - triangle_median/2
    )
    C = (
        triangle_center_x - (np.sqrt(3)*triangle_median)/2,
        triangle_center_y - triangle_median/2
    )

    plt.plot([A[0],B[0]], [A[1], B[1]], 'k', linewidth=0.5)
    plt.plot([A[0],C[0]], [A[1], C[1]], 'k', linewidth=0.5)
    plt.plot([B[0],C[0]], [B[1], C[1]], 'k', linewidth=0.5)

    beta_angle = np.pi

    A = (
        triangle_center_x * beta_angle,
        (triangle_center_y + triangle_median) * beta_angle
    )
    B = (
        (triangle_center_x + (np.sqrt(3)*triangle_median)/2) * beta_angle,
        (triangle_center_y - triangle_median/2) * beta_angle
    )
    C = (
        (triangle_center_x - (np.sqrt(3)*triangle_median)/2) * beta_angle,
        (triangle_center_y - triangle_median/2) * beta_angle
    )

    plt.plot([A[0],B[0]], [A[1], B[1]], 'k', linewidth=0.5)
    plt.plot([A[0],C[0]], [A[1], C[1]], 'k', linewidth=0.5)
    plt.plot([B[0],C[0]], [B[1], C[1]], 'k', linewidth=0.5)
plt.axis('square')
plt.axis('off')
# plt.title(f'n={n_points_circle}')
my_date = datetime.now().strftime('%y%m%d_%H_%M_%S')
plt.savefig(f'{my_date}_triangle_{n_points_circle}_median_{triangle_median}_no_rotation.svg')
plt.show()
```
