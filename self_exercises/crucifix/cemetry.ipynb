```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
cross_width = 20
cross_height = 1.618 * cross_width
cross_head_y = 0.146 * cross_width
cross_chest_y = 0.382 * cross_width
cross_legs_y = 1.118 * cross_width

cross_head_x = 0.382 * cross_width
cross_legs_x = 0.25 * cross_width


w_margin = 4
h_margin = 4

rows, cols = 60, 40

gryadka = 5

my_row = np.random.randint(5, rows)
my_col = np.random.randint(5, cols)

while my_row % gryadka == 0 or my_col % gryadka == 0:
    my_row = np.random.randint(5, rows)
    my_col = np.random.randint(5, cols)

print(f'Your row is {my_row}, your col is: {my_col}.')

for r_idx, row in enumerate(range(rows)):
    for c_idx, col in enumerate(range(cols)):

        if r_idx % gryadka == 0  or c_idx % gryadka  == 0:
            continue
        if r_idx == my_row and c_idx == my_col:
            continue

        top_center = (
            r_idx * (cross_width + 2*w_margin),
            c_idx * (cross_height + 2*h_margin)
        )

        plt.plot(
            [top_center[0],top_center[0]],
            [top_center[1],top_center[1]-cross_height],
            'k',
            linewidth=0.3
        )
        
        plt.plot(
            [top_center[0]-0.5*cross_width,top_center[0]+0.5*cross_width],
            [top_center[1]-cross_chest_y, top_center[1]-cross_chest_y],
            'k',
            linewidth=0.3
        )

        plt.plot(
            [top_center[0]-0.5*cross_head_x,top_center[0]+0.5*cross_head_x],
            [top_center[1]-cross_head_y, top_center[1]-cross_head_y],
            'k',
            linewidth=0.3
        )

        plt.plot(
            [top_center[0] - 0.5*cross_legs_x, top_center[0] + 0.5*cross_legs_x],
            [top_center[1] - cross_legs_y + (0.5*cross_legs_x)/np.sqrt(2), top_center[1] - cross_legs_y - (0.5*cross_legs_x)/np.sqrt(2)],
            'k',
            linewidth=0.3
        )


        # plt.plot([top_center[0]-2, top_center[0]+2], [top_center[1]-3, top_center[1]-3], 'k', linewidth=0.3)



plt.axis('off')
plt.savefig('crucifix.svg')
plt.show()
```
