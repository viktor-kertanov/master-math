```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
x_dot = [1,10]
y_dot = [2,3]

tf = 2


# display(Math(sym.latex(sym.sympify(tau))))

x = []
y = []

t_choice = np.linspace(0, 2, 1000)
t_random = np.random.choice(t_choice, 1000)

for t_idx, t in enumerate(t_random, start=1):
    tau = t / tf
    polynom_factor = (15*tau**4 - 6*tau**5 - 10*tau**3)
    print(f"Polynom_factor when t = {t:.2f}: {polynom_factor:.2f}")
    xt = x_dot[0] + (x_dot[1] - x_dot[0])*polynom_factor
    yt = y_dot[0] + (y_dot[1] - y_dot[0])*polynom_factor
    x.append(xt)
    y.append(yt)
    # print(f"Step #{t_idx}, t={t:.2f}, xt={xt:.1f}, yt={yt:.1f}")



plt.plot(x, y)

plt.axis()

plt.show()
```
