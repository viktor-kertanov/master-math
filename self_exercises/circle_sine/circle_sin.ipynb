```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 4
h = 0
k = 0

axlim = r + np.max(abs(h), abs(k))+1
x = np.linspace(-axlim, axlim, 1000)
X, Y = np.meshgrid(x, x)

Fxy = (X - h)**2 + (Y - k)**2 - r**2

plt.contour(X, Y, Fxy, 0, colors='k')

plt.gca().set_aspect('equal')
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 1000)
x = np.cos(angles)
y = np.sin(angles)

plt.plot(x, y, 'k')
plt.axis('square')
plt.show()
```

```python
for r in np.linspace(.1, 5, 150):
    angles = np.linspace((np.random.rand()-.4), 2*np.pi*(1-np.random.rand()/2.2), 1000)
    for _ in range(np.random.randint(2,4)):
        plt.polar(angles, r+0.4*np.sin(np.pi*1.2*angles*r)*np.exp(-1.75*r/np.pi)+np.random.rand()/50, color='k', linewidth=.5)

plt.axis('off')
plt.gcf().set_size_inches(7,5)
plt.savefig('rose_polar_chaos.svg')
plt.show()
angles = np.linspace(-np.abs(np.random.rand()*.001), 2*np.pi+np.abs(np.random.rand()*.001), 1000)
for _ in range(np.random.randint(3,5)):
        plt.polar(angles, r+0.4*np.sin(np.pi*1.2*angles*r)*np.exp(-np.pi), color='b', linewidth=2)

plt.axis('off')
plt.gcf().set_size_inches(7,5)
plt.savefig('rose_polar_contour.svg')
plt.show()
```
