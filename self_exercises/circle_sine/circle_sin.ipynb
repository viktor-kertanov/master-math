```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 4
h = 0
k = 0

axlim = r + np.max(abs(h), abs(k))+1
x = np.linspace(-axlim, axlim, 1000)
X, Y = np.meshgrid(x, x)

Fxy = (X - h)**2 + (Y - k)**2 - r**2

plt.contour(X, Y, Fxy, 0, colors='k')

plt.gca().set_aspect('equal')
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 1000)
x = np.cos(angles)
y = np.sin(angles)

plt.plot(x, y, 'k')
plt.axis('square')
plt.show()
```
