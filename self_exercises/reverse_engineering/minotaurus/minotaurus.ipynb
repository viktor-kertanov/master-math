```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

angles = np.linspace(0, 2*np.pi, 100)



for i in np.arange(100, 200):
    x = np.cos(angles)*i**(1/5)-i**(1/5)
    y = np.sin(angles)*i**(1/5)

    plt.plot(x, y, '-k', linewidth=1)

for i in np.arange(40, 100, 10):
    x = np.cos(angles)*i**(1/5)-i**(1/5)
    y = np.sin(angles)*i**(1/5)

    plt.plot(x, y, '-k', linewidth=1)

# for i in np.arange(10, 20):
#     x = np.cos(angles)*i**(1/10)-i
#     y = np.sin(angles)*i**(1/10)

#     plt.plot(x, y, '-k', linewidth=1)
# x = np.cos(angles)-1
# y = np.sin(angles)
# plt.plot(x, y, '-k', linewidth=1)

plt.axis('square')
# plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(8.3, 5.8) # A5 in inches
plt.show()
```
