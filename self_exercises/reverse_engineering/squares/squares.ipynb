```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
x = 0
y = 0

a_max = 12
a_min = 8
a_step = 1

ax = plt.axes()
ax.set_facecolor('black')

# main squares
line_color = 'w'
linewidth = 1
for a in np.arange(a_min, a_max+a_step, a_step):
    plt.plot([x-a/2, x-a/2, x+a/2, x+a/2, x-a/2], [y-a/2, y+a/2, y+a/2, y-a/2, y-a/2], line_color, linewidth=linewidth)

# four сornersquares
b = (a_max-a_min)*(a_step/2)/2+a_step*1.5
signs = [(1, 1), (-1, 1), (1, -1), (-1, -1)]
for i in signs:
    plt.plot(
        np.array([x - a_max/2, x - a_max/2 + b, x - a_max/2 + b])*i[0],
        np.array([y + a_max/2 - b, y + a_max/2-b, y+a_max/2])*i[1],
        line_color,
        linewidth=linewidth)


# inner square
a_max = 7.5
a_min = 1.5
a_step = .75

line_color = 'w'
for a in np.arange(a_min, a_max+a_step, a_step):
    plt.plot([x-a/2, x-a/2, x+a/2, x+a/2, x-a/2], [y-a/2, y+a/2, y+a/2, y-a/2, y-a/2], line_color, linewidth=linewidth)

# four сornersquares
b = (a_max-a_min)*(a_step/2)/2+ a_step+.25
signs = [(1, 1), (-1, 1), (1, -1), (-1, -1)]
for i in signs:
    plt.plot(
        np.array([x - a_max/2, x - a_max/2 + b, x - a_max/2 + b])*i[0],
        np.array([y + a_max/2 - b, y + a_max/2-b, y+a_max/2])*i[1],
        line_color,
        linewidth=linewidth)


plt.gcf().set_size_inches(8, 8)
plt.gca().set_aspect('equal')
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])
plt.xticks([])
plt.yticks([])
plt.show()
```
