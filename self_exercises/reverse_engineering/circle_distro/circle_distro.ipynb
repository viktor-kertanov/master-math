```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean_x = 0
sigma = np.pi
n = 30
random_nums = np.random.randn(n)*sigma
random_nums -= np.mean(random_nums)

circle_y = np.sin(random_nums)

y = np.zeros(n)

plt.plot(random_nums, y, '.k')
plt.plot(random_nums, circle_y, '.k')

# plt.gca().set_aspect('equal')
plt.show()
```

```python
mean = np.pi/4
sigma = np.pi
n = 30
angles = np.random.randn(n)*sigma + mean
# angles = np.random.exponential(scale=1, size=n) + mean
# angles = angles[(angles > 0) & (angles < np.pi/2)]

# angles = np.linspace(0, np.pi/2, 100)
# x = np.abs(0 - np.cos(angles))
x = np.cos(0-angles)
y = np.sin(angles)

coords = zip(x, y)
r = 1
for coord in coords:
    random_param = np.random.rand()*0.02
    plt.plot([coord[0], coord[0]], [coord[1]*(r-random_param), -coord[1]*(r-random_param)], '-k', linewidth=1 )
    plt.plot([-coord[0], -coord[0]], [coord[1]*(r-random_param), -coord[1]*(r-random_param)], '-k', linewidth=1 )

plt.gca().set_aspect('equal')
plt.axis('square')
plt.gcf().set_size_inches(8, 8)
plt.show()
```

```python
rows = 3
cols_in_row=5
fig, ax = plt.subplots(rows, cols_in_row, figsize=(15, 15))
fig.tight_layout()
row_idx, col_idx = np.indices((rows, cols_in_row))

mean = np.pi/4
sigma = np.pi

for idx in np.arange(rows*cols_in_row):
    n = np.random.randint(30, 100)
    angles = np.random.randn(n)*sigma + mean
    # angles = np.random.exponential(scale=1, size=n) + mean
    angles = angles[(angles > 0) & (angles < np.pi/2)]

    # angles = np.linspace(0, np.pi/2, 100)
    # x = np.abs(0 - np.cos(angles))
    x = np.cos(0-angles)
    y = np.sin(angles)

    coords = zip(x, y)
    r = 1

    cur_plt = ax[row_idx.ravel()[idx], col_idx.ravel()[idx]]
    for coord in coords:
        random_param = np.random.rand()*0.02
        cur_plt.plot([coord[0], coord[0]], [coord[1]*(r-random_param), -coord[1]*(r-random_param)], '-k', linewidth=1 )
        cur_plt.plot([-coord[0], -coord[0]], [coord[1]*(r-random_param), -coord[1]*(r-random_param)], '-k', linewidth=1 )

    cur_plt.axis('square')
    cur_plt.axis([-r, r, -r, r])
    cur_plt.axis('off')

# plt.gcf().set_size_inches(8, 8)
plt.subplots_adjust(wspace=0.1, hspace=0.2)
plt.savefig('circles_distro.svg', dpi=100)
plt.show()
```

```python
mean = np.pi/4
sigma = np.pi
n = 30
# angles = np.random.exponential(scale=2, size=n)/(np.pi/1.5)
angles = np.linspace(0, np.pi/2, 10)
x = np.cos(angles)
y = np.sin(angles)

coords = zip(x, y)
r = 6
for idx, coord in enumerate(coords):
    random_param=0
    dots_x = np.linspace(-coord[0]*r, coord[0]*r, 10000)
    dots_y = np.sin(5*r*dots_x)/5 + coord[1]*r
    plt.plot(dots_x, dots_y, 'k-', linewidth=1)
    plt.plot(dots_x, -dots_y, 'k-', linewidth=1)
    plt.plot([-coord[0]*r, coord[0]*r], [coord[1]*(r-random_param), coord[1]*(r-random_param)], '-b', linewidth=1 )
    plt.plot([-coord[0]*r, coord[0]*r], [-coord[1]*(r-random_param), -coord[1]*(r-random_param)], '-b', linewidth=1 )

plt.axis([-r, r, -r, r])
plt.axis('square')
plt.axis('off')
plt.gcf().set_size_inches(8, 8)
# plt.savefig('funny_error_6.svg', dpi=100)
plt.show()
```
