```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = .3

for a in range(1, 5):
    for b in range(1, 5):
        x = np.linspace(-a, a, 100)
        y = np.linspace(-b, b, 100)
        X, Y = np.meshgrid(x, y)
        Fxy = (np.abs(X)/a)**(2*a/r) + (np.abs(Y)/b)**(2*b/r) - 1

        plt.contour(X, Y, Fxy, linewidths=1, colors='k', levels=0)
plt.axis([-a-1, a+1, -b-1, b+1])
plt.show()
```

```python
for r in np.arange(.1, 10, 10):
    color = (np.random.rand(), np.random.rand(), np.random.rand())
    for a in range(1, 10):
        for b in range(1, 10):
            x = np.linspace(-a, a, 100)
            y = np.linspace(-b, b, 100)
            X, Y = np.meshgrid(x, y)
            Fxy = (np.abs(X)/a)**(2*a/r) + (np.abs(Y)/b)**(2*b/r) - r
            plt.contour(X, Y, Fxy, linewidths=1, colors=color, levels=0)

plt.axis([-a-1, a+1, -b-1, b+1])
plt.show()
```

```python
r = 10
a = 3
b = 5

for r in np.logspace(-1, 5, 40):
    x = np.linspace(-a, a, 100)
    y = np.linspace(-b, b, 100)
    X, Y = np.meshgrid(x, y)
    Fxy = (np.abs(X)/a)**(2*a/r) + (np.abs(Y)/b)**(2*b/r) - 1
    color = np.round(np.random.rand(4),2)
    try:
        plt.contour(X, Y, Fxy, linewidths=1, colors=color, levels=0)
    except ValueError:
        plt.contour(X, Y, Fxy, linewidths=1, levels=0)
        print(color)

plt.axis([-a*1.1, a*1.1, -b*1.1, b*1.1])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(15, 10)
plt.axis('off')
plt.show()
```
