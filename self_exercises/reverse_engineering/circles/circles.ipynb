```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
import sympy as sym
```

```python
angles = np.linspace(0, 2*np.pi, 100)

for i in np.arange(0.05, 1, .05):
    x = np.cos(angles)*i
    y = np.sin(angles)*i

    plt.plot(x, y, '-k', linewidth=1)
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(8.3, 5.8) # A5 in inches
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 100)

for i in np.arange(.1, 1, .1):
    x = np.cos(angles)*i-i
    y = np.sin(angles)*i

    plt.plot(x, y, '-k', linewidth=1)
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(8.3, 5.8) # A5 in inches
plt.show()
```

```python
angles = np.linspace(0, 2*np.pi, 1000)

ax = plt.axes()
ax.set_facecolor('black')

for i in np.logspace(np.log10(1.6), np.log10(2*np.pi), 15):
    x = np.cos(angles)*i-i
    y = np.sin(angles)*i
    filter_array = [True if el > 0.1 else False for el in y]

    plt.plot(x[filter_array], y[filter_array], 'w', linewidth=2)

    x = np.cos(angles)*i+i - 12.55
    y = np.sin(angles)*i
    filter_array = [True if el < -0.1 else False for el in y]
    plt.plot(x[filter_array], y[filter_array], 'w', linewidth=2)

plt.gca().set_aspect('equal')
plt.gca().axes.xaxis.set_ticklabels([])
plt.gca().axes.yaxis.set_ticklabels([])
plt.xticks([])
plt.yticks([])
plt.gcf().set_size_inches(11.7, 8.3) # A5 in inches

plt.savefig('concentric.svg', dpi=100)
plt.show()
```
