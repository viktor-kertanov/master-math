```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
x_s = np.arange(n)
y_s = np.arange(n)
scl = 10

thetas = [0, np.pi/8, -np.pi/4, -np.deg2rad(-67.5)]
colors = ['k', 'r', 'g', 'b']
colors = ['k', 'k', 'k', 'k']
for t_idx, theta in enumerate(thetas):
    for x in x_s:
        x_arr = np.array([x, x])*scl
        y_arr = np.array([y_s[0], y_s[-1]])*scl

        x = np.cos(theta)*x_arr - np.sin(theta)*y_arr
        y = np.sin(theta)*x_arr + np.cos(theta)*y_arr
        plt.plot(x, y, colors[t_idx], linewidth=1)
    for y in y_s:
        x_arr = np.array([x_s[0], x_s[-1]])*scl
        y_arr = np.array([y, y])*scl

        x = np.cos(theta)*x_arr - np.sin(theta)*y_arr
        y = np.sin(theta)*x_arr + np.cos(theta)*y_arr
        plt.plot(x, y, colors[t_idx], linewidth=1)

plt.axis([500, 600, 250, 350])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(13, 10)
plt.show()
```

```python
# another approach to use a y = kx+b formula to solve the limits thing
n = 100

b_s = np.arange(100)
k_s = np.array([0, np.pi/8, -np.pi/4, -np.deg2rad(-67.5)])
tangents = np.tan(k_s)

x_min, x_max = 0, 100


for k in tangents:
    for b in b_s:
        plt.plot([x_min, x_max],[k*x_min+b, k*x_max+b], 'k', linewidth=1)
        plt.plot([k*x_min+b, k*x_max+b], [x_min, x_max], 'k', linewidth=1)


plt.axis([0, 50, 0, 50])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(13, 10)
plt.show() 
```

```python
# another approach to use a y = kx+b formula to solve the limits thing
thetas = np.array([22.5, -45, -67.5])
thetas_1 = thetas+90
thetas = np.concatenate((thetas, thetas_1))
print(thetas)
thetas_rad = np.tan(np.deg2rad(thetas))
print(thetas_rad)
# col = ['c', 'r', 'g', 'b','c', 'r', 'g', 'b']
b_delta_main = 1
linewidth=1

canvas_min_x = 0
canvas_max_x = 500
canvas_min_y = 0
canvas_max_y = 500

# ax = plt.axes()
# ax.set_facecolor('black')

#plotting for 0
x = np.arange(canvas_min_x, canvas_max_x, b_delta_main)
for b in np.arange(canvas_min_x, canvas_max_x, b_delta_main):
    plt.plot(x, x*0+b, 'b', linewidth=linewidth)
    plt.plot(x*0+b, x, 'b', linewidth=linewidth)

for a_idx, a in enumerate(thetas_rad):
    critical_points = [(canvas_min_x, canvas_max_y), (canvas_max_x, canvas_min_y), (canvas_max_x, canvas_max_y), (canvas_min_x, canvas_min_y)] 
    b_coef = [xy[1]-xy[0]*a for xy in critical_points]
    min_b = min(b_coef)
    max_b = max(b_coef)
    b_delta = np.sqrt(b_delta_main + a**2)
    x = np.arange(canvas_min_x, canvas_max_x, b_delta_main)
    for b in np.arange(min_b, max_b, b_delta):
        plt.plot(x, a*x+b, 'b', linewidth=linewidth)

scl = 7
plt.axis([canvas_min_x/scl, canvas_max_x/scl, canvas_min_y/scl, canvas_max_y/scl])
plt.axis('off')
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(15,15)
plt.savefig('greed_blue.svg', dpi=1000)
# plt.grid()
plt.show() 
```
