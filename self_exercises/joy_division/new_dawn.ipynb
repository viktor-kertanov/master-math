```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=30


y_filtered = np.zeros(n)

fig, ax = plt.subplots(figsize=(5,5))
for i in range(80):
    x = np.linspace(-3*(1+np.random.rand()/100), 3*(1+np.random.rand()/100), n)
    y = np.abs(np.random.randn(n)*np.exp(-np.abs(2*x)))/2+i/10
    y_filtered = np.where(y>y_filtered, y, y_filtered+np.random.rand()*0.04)
    
    randomizer = np.random.rand()
    for _ in range(3, 7):
        if randomizer > .1:
            plt.plot(x, y_filtered,  'k', linewidth=1)
        else:
            plt.plot(x, y, 'k', linewidth=1)

# plt.gca().set_aspect('equal')
# plt.gcf().set_size_inches(7, 5)
# plt.ylim((0, 25))
plt.axis('off')
plt.savefig('new_dawn_fades.svg', dpi=100)
plt.show()
```

```python
def new_dawn_fades():
    # define the mathematical function to plot
    t = np.linspace(0, 6*np.pi, 1000)
    x = np.sin(t) * np.exp(-0.1*t)
    y = np.cos(t) * np.exp(-0.1*t)

    # plot the function using matplotlib
    fig, ax = plt.subplots(figsize=(8,8))
    ax.plot(x, y, color='white', linewidth=3)

    # add the vertical and horizontal lines
    ax.axvline(0, color='white', linewidth=3)
    ax.axhline(0, color='white', linewidth=3)

    # add the square and circle
    square = plt.Rectangle((-1, -1), 2, 2, color='white', linewidth=3, fill=False)
    ax.add_patch(square)
    circle = plt.Circle((0, 0), 0.5, color='white', linewidth=3, fill=False)
    ax.add_artist(circle)

    # set the axis limits and background color
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_facecolor('black')

    # remove the axis ticks
    ax.set_xticks([])
    ax.set_yticks([])

    # show the plot
    plt.show()

new_dawn_fades()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def new_dawn_fades():
    # define the mathematical function to plot
    t = np.linspace(0, 10*np.pi, 1000)
    x = np.sin(t) * np.exp(-0.1*t)
    y = np.cos(t) * np.exp(-0.1*t)

    # plot the function using matplotlib
    fig, ax = plt.subplots(figsize=(8,8))
    ax.plot(x, y, color='white', linewidth=3)

    # add the vertical and horizontal lines
    ax.axvline(0, color='white', linewidth=3)
    ax.axhline(0, color='white', linewidth=3)

    # add the square and circle
    square = plt.Rectangle((-1, -1), 2, 2, color='white', linewidth=3, fill=False)
    ax.add_patch(square)
    circle = plt.Circle((0, 0), 0.5, color='white', linewidth=3, fill=False)
    ax.add_artist(circle)

    # add the two intersecting lines
    line1 = plt.Line2D([-0.5, 0.5], [1, -1], color='white', linewidth=3)
    line2 = plt.Line2D([-0.5, 0.5], [-1, 1], color='white', linewidth=3)
    ax.add_artist(line1)
    ax.add_artist(line2)

    # set the axis limits and background color
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)
    ax.set_facecolor('black')

    # remove the axis ticks
    ax.set_xticks([])
    ax.set_yticks([])

    # show the plot
    plt.show()

new_dawn_fades()
```
