```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_deltas = np.linspace(-10, 10, 100)
x = np.linspace(-10*np.pi, 10*np.pi, 10000)
for y_delta in y_deltas:
    for phase in np.linspace(1,3,3):
        plt.plot(x, np.sin(x)*phase*0.1+y_delta, 'k', linewidth=0.5)

# plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
y_deltas = np.linspace(-10, 10, 150)
for y_delta in y_deltas:
    for phase in range(-12,12, 1):
        if phase == -1: 
            continue
        x = np.linspace(phase*np.pi, (phase+1)*np.pi, 2000)
        plt.plot(x[:-1], np.sin(x[:-1])*(phase+1)*0.25+y_delta, 'k', linewidth=0.25)

# plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('sine_axi.svg')
plt.show()
```

```python
y_deltas = np.linspace(-10, 10, 200)
x = np.linspace(-10*np.pi, 10*np.pi, 10000)
for y_delta in y_deltas:
    plt.plot(x, np.sin(x)*0.2+y_delta, 'k', linewidth=0.5)

# plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('calm_sine.svg')
plt.show()
```

```python
y_deltas = np.linspace(-10, 10, 150)
for y_delta in y_deltas:
    for phase in range(-12,12, 1):
        if phase == -1: 
            continue
        x = np.linspace(phase*np.pi, (phase+1)*np.pi, 2000)
        plt.plot(x[:-1], np.sin(x[:-1])*(phase+1)*np.random.rand()+y_delta, 'k', linewidth=0.25)

# plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('sine_axi_random.svg')
plt.show()
```
