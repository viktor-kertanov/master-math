```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
sine_stretch = .3

x_start = 1*np.pi
x_stop = 6*np.pi

microdose = np.pi/100

x = np.linspace(x_start, x_stop, n)
turq_col = (9/255, 60/255, 80/255)
dark_red = (100/255, 0, 0)

thick_part_num = 15
y_range = 0
for k in range(50):
    x_red_init = np.random.choice(x, 1)

    for i in range(thick_part_num):
        # moving left-right thick color lines in phases
        if (k // 10) % 2 == 0:
            shift_direction = -1
        else:
            shift_direction = 1
        
        # y values with a shift along x-axis
        y = np.sin(x+microdose*i*1.2 + shift_direction*microdose*k*10)*sine_stretch - microdose*i + y_range*k

        # max_indexes = np.argmax(y)
        # min_indexes = np.argmax(-y)

        if i == 0:
            y_max = np.max(y)
        if i == thick_part_num-1:
            y_min = np.min(y)
            y_range = y_max - y_min
        
        plt.plot(x, y, linewidth=.8, color=turq_col)

        x_red = np.linspace(x_red_init-microdose*2*i, x_red_init+microdose*6*i)
        y_red = np.sin(x_red+microdose*i + shift_direction*microdose*k*10)*sine_stretch - microdose*i + y_range*k
        plt.plot(x_red, y_red, linewidth=1, color=dark_red, alpha=1)

plt.xlim([x_start, x_stop])
# plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(15, 11)
plt.axis('off')
plt.savefig('zhanna_sine.svg', dpi=100)
plt.savefig('zhanna_sine.pdf', dpi=100)
plt.savefig('zhanna_sine.jpeg', dpi=1000)
plt.show()
```
