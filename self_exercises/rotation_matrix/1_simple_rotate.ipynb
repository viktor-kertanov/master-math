```python
x = np.linspace(0, 5, 100)
fx = x*0

for theta_deg in np.linspace(0, 360, 5)[:-1]:
    theta = np.deg2rad(theta_deg)
    sin_theta, cos_theta = np.sin(theta), np.cos(theta)
    x_r = cos_theta*x - sin_theta*fx
    fx_r = sin_theta*x + cos_theta*fx

    plt.plot(x_r, fx_r, color=(theta_deg/360, theta_deg/360, theta_deg/360), label=f'$\\theta$={theta_deg}')

plt.axis('square')
plt.legend()
plt.show()
```

Succsess witht the sinus function below!
1. For Axi Draw it is possible to create many svgs with a cumulative number of layers (360 svgs: on the first svg only one sinus function, on the second -- 2 lines etc.)
2. Merge all the layers into one svg
3. Draw everything: first lines will be more bright and the last will be very gentle

```python
# Below is "pure" sine function, without any additional random variables
x = np.linspace(0, 10*np.pi, 100)
fx = np.sin(x)

for theta_deg in np.linspace(0, 360, 360)[:-1]:
    theta = np.deg2rad(theta_deg)
    sin_theta, cos_theta = np.sin(theta), np.cos(theta)
    x_r = cos_theta*x - sin_theta*fx
    fx_r = sin_theta*x + cos_theta*fx

    plt.plot(
        x_r,
        fx_r,
        color=(theta_deg/360, theta_deg/360, theta_deg/360),
        label=f'$\\theta$={theta_deg}',
        linewidth=1
    )
plt.gcf().set_size_inches(15, 10)
plt.axis('square')
plt.axis('off')
# plt.legend()
plt.show()
```

```python
for theta_deg in np.linspace(0, 360, 360)[:-1]:
    x = np.linspace(np.random.rand()*2*np.pi, (14-np.random.rand()*theta_deg/10)*np.pi, 10000) # instead of zero: np.random.rand()*4*np.pi
    fx = np.sin(x)
    theta = np.deg2rad(theta_deg)
    sin_theta, cos_theta = np.sin(theta), np.cos(theta)
    x_r = cos_theta*x - sin_theta*fx
    fx_r = sin_theta*x + cos_theta*fx

    plt.plot(
        x_r,
        fx_r,
        color='k',
        label=f'$\\theta$={theta_deg}',
        linewidth=0.5
    )
plt.gcf().set_size_inches(15, 10)
plt.axis('square')
plt.axis('off')
# plt.legend()
plt.savefig('random_sin_rotated.svg')
plt.show()
```

## For axidraw: instead of changing color we'll have overlapping lines to have the same effect

```python
# Below is "pure" sine function, without any additional random variables
max_line_overlap = 15

t = 360
thetas_deg = np.linspace(0, 360, t)[:-1]

overlap_segments = int(np.ceil(t/max_line_overlap))
overlap_idx = 0

for theta_idx, theta_deg in enumerate(thetas_deg, start=1):
    x = np.linspace(np.random.rand()*np.pi/8+1.5*np.pi, 10*np.pi+np.random.rand()*np.pi/8, 100)
    fx = np.sin(x)*0.8
    theta = np.deg2rad(theta_deg)
    sin_theta, cos_theta = np.sin(theta), np.cos(theta)
    x_r = cos_theta*x - sin_theta*fx
    fx_r = sin_theta*x + cos_theta*fx

    overlaps = max_line_overlap - int(theta_idx//overlap_segments)
    
    while overlap_idx <= overlaps:
        plt.plot(
            x_r,
            fx_r,
            color='k',
            label=f'$\\theta$={theta_deg}',
            linewidth=0.5
        )
        overlap_idx+=1
    
    overlap_idx=0
plt.gcf().set_size_inches(5.28*2, 8.26*2)
plt.axis('square')
plt.axis('off')
# plt.legend()
plt.savefig('sin_rotate_overlap.svg', transparent=True)
plt.show()
```
