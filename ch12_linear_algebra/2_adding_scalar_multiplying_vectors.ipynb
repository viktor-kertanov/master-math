### **Adding and Subtracting vectors and multiplying a vector by a scalar**

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
v = np.array([.5, 1])
scalars = np.array([1, -.5, 2, .5])
```

```python
for idx, s in enumerate(scalars, start=1):
    scaled_vector = v * s
    plt.plot([0, scaled_vector[0]], [0, scaled_vector[1]],'o-', linewidth=7-idx, markersize=8-idx, alpha=(8-idx)/8, label='$\lambda = %g$' % s)

plt.axis('square')
plt.axis([-3, 3, -3, 3])
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
# Vectors
v1 = np.array([-1, 2])
v2 = np.array([1, 1])

# Different operations options
v3_a = v1 + v2
v3_b = np.add(v1, v2)
v3_c = np.zeros(2)

for i in range(2):
    v3_c[i] = v1[i] + v2[i]

print(v3_a, v3_b, v3_c)
```

### Confirming that the geometric interpretation of vector sum is correct if we do it algebraically 

```python
v1 = np.array([-1, 2])
v2 = np.array([1, 1])

v3 = np.add(v1, v2)

# v1
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='$v1$')
plt.plot([0, v1[0]] + v2[0], [0, v1[1]]+v2[1], 'b--', linewidth=2, alpha=0.8, label='$v1\'$')
# v2
plt.plot([0, v2[0]] + v1[0], [0, v2[1]] + v1[1], 'g--', alpha=0.8, label='$v2\'$')
plt.plot([0, v2[0]], [0, v2[1]], 'g', alpha=0.8, linewidth=2, label='$v2$')

# v3 = v1 + v2
plt.plot([0,v3[0]], [0, v3[1]], 'm', linewidth=2, label='$v1+v2$')

display(Math('%s + %s = %s' % (sym.latex(sym.sympify(v1)), sym.latex(sym.sympify(v2)), sym.latex(sym.sympify(v3)))))

# plot
plt.axis('square')
plt.axis([-2, 2, -2, 4])
plt.grid()
plt.legend(loc='lower left', fontsize=6)


plt.show()
```
