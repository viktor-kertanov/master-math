```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display, Math
```

```python
row_vector = np.array([4, 2])
col_vector = np.array([ [-2], [3]])

print(row_vector)
print(col_vector)
```

```python
# in latex coding
display(Math(sym.latex(sym.sympify(row_vector))))
display(Math(sym.latex(sym.sympify(col_vector))))
```

```python
# when plotting, we start from "standard position", i.e. (0,0)

plt.plot([0, row_vector[0]], [0, row_vector[1]], 'r', label='row vector')
plt.plot([0, col_vector[0][0]], [0, col_vector[1][0]], 'b', label='col vector')


plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.legend()
plt.show()
```

```python
# exercise

M = np.array([[3, 0, -4], [-1, 1, 3]])

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection="3d")

v1=[3, 0, -4]
v2=[-1, 1, 3]

ax.plot3D([-4, 4], [0,0], [0,0], 'k--', alpha = 0.5)
ax.plot3D([0, 0], [-4,4], [0,0], 'k--', alpha = 0.5)
ax.plot3D([0, 0], [0,0], [-4,4], 'k--', alpha = 0.5)
ax.plot3D([0, 3], [0,0], [0,-4], 'b')
ax.plot3D([0, -1], [0,1], [0,3], 'r')

plt.ylabel('y')
plt.xlabel('x')
ax.set_xlim3d(-5,5)
ax.set_ylim3d(-5,5)
ax.set_zlim3d(-5,5)

plt.gcf().set_size_inches(18, 10)
plt.show()
```

```python
# exercise, more comprehensive technique

M = np.array([[3, 0, -4], [-1, 1, 3]])
R = np.random.rand(1000,3)*8-4

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection="3d")

v1=[3, 0, -4]
v2=[-1, 1, 3]

#axes
ax.plot3D([-4, 4], [0,0], [0,0], 'k--', alpha = 0.5, linewidth=0.5)
ax.plot3D([0, 0], [-4,4], [0,0], 'k--', alpha = 0.5, linewidth=0.5)
ax.plot3D([0, 0], [0,0], [-4,4], 'k--', alpha = 0.5, linewidth=0.5)

#vectors
for vector in R:
    ax.plot3D([0, vector[0]], [0, vector[1]], [0, vector[2]], '', alpha=0.3)

plt.ylabel('y')
plt.xlabel('x')
plt.axis('on')
plt.gcf().set_size_inches(18, 10)
plt.show()
```
