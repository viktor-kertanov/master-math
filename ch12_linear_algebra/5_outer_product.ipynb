### __The Outer Product__

The outer product, unlike the dot product (or "inner product" ), can be calculated for the vectors of different length.

We take the column-vector and a row vector and the result is an entire matrix, where columns are repeated and rows are repeated:

<br>

\begin{equation}
\begin{bmatrix}
1 \\
0 \\
2 \\
5
\end{bmatrix}
\times
\begin{bmatrix}
a & b & c & d
\end{bmatrix}
=
\begin{bmatrix}
1a & 1b & 1c & 1d \\
0a & 0b & 0c & 0d \\
2a & 2b & 2c & 2d \\
5a & 5b & 5c & 5d \\
\end{bmatrix}
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
v1 = np.random.randn(50)
v2 = np.random.randn(80)

out_product = np.outer(v1, v2)

plt.imshow(out_product)

plt.show()
```

#### __Exercise__
We'll have to test if the outer product is __matrix commutative__ and __scalar-multiplication commutative__:

$$ a \times b = b \times a$$

Matrix commutative:
$$ \text{np.outer(v1, v2) = np.outer(v2, v1)} $$

Scalar-multiplication commutative:
$$cvw^T=vcw^T=vw^Tc $$

```python
v1_num = 10
v2_num = 7

v1 = np.random.randn(v1_num)
v2 = np.random.randn(v2_num)

outer_1 = np.outer(v1, v2)
outer_2 = np.outer(v2, v1)

# Matrix commutative
plt.subplot(131)
plt.imshow(outer_1)
plt.title('Outer product (v1, v2)', fontsize=10)

plt.subplot(132)
plt.imshow(outer_2.T)
plt.title('Outer product $(v2, v1)^T$', fontsize=10)

plt.subplot(133)
plt.imshow(outer_2)
plt.title('Outer product (v2, v1)', fontsize=10)


plt.show()

display(Math(sym.latex(sym.sympify(outer_1))))
display(Math(sym.latex(sym.sympify(outer_2.T))))

```

### __Now for scalar commutativity__

```python
scalar = 4
scalar_0 = scalar*np.outer(v1, v2)
scalar_1 = np.outer(v1*scalar, v2)
scalar_2 = np.outer(v1, v2)*scalar
scalar_3 = np.outer(v1, v2)*scalar

display(Math(sym.latex(sym.sympify(scalar_0))))
display(Math(sym.latex(sym.sympify(scalar_1))))
display(Math(sym.latex(sym.sympify(scalar_2))))
display(Math(sym.latex(sym.sympify(scalar_3))))
display(Math(sym.latex(sym.sympify(scalar_1-scalar_2))))
```

#### __Solution__: as a result we have the same matrices, i.e. outer product is a scalar-commutative operation
