```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

### __Visualizing matrix-vector multiplication__

There's an interesting geometric interpretations of matrix-vector multiplication. <br>
We need this part to build some intuition about __eigenvectors (собственные вектора)__. <br>
Also we'll be able to build some more beautiful art. <br>

```python
M = np.array([
    [.5, 1],
    [1, .5 ]
    ])

v = np.array(
    [1, 3]
    )

# v = np.random.randn(2)

Mv = M @ v

display(Math(sym.latex(sym.sympify(Mv))))

plt.plot([0, v[0]], [0, v[1]], 'b', linewidth=2, label='v')
plt.plot([0, Mv[0]], [0, Mv[1]], 'r--', linewidth=2, label='Mv')

plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.legend()
plt.show()
```

So when we take a vector and multiply it by a matrix, what we get is:
1. The vector __shrinks or extends__, it changes size;
2. The vector __rotates__.

<br>

Below we see that if the vector is __[1, 1]__ , than the Mv is only changes in size but is not rotating. It is kinda the same a multiplying by a scalar. The matrix acts like a scalar.
If we take the vector __[1, -1]__ it will change direction, going opposite side.


```python
M = np.array([
    [.5, 1],
    [1, .5 ]
    ])

v = np.array(
    [1, 1]
    )

# v = np.random.randn(2)

Mv = M @ v

display(Math(sym.latex(sym.sympify(Mv))))

plt.plot([0, v[0]], [0, v[1]], 'b', linewidth=2, label='v')
plt.plot([0, Mv[0]], [0, Mv[1]], 'r--', linewidth=2, label='Mv')

plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.legend()
plt.show()
```

```python
# Exercise

M = np.array([
    [.5, 1],
    [1, .5]
])

n = 50

r_1 = np.linspace(0.5, 1, int(n/2))
g_1 = np.linspace(0, 0.5, int(n/2))
b_1 = np.linspace(0.5, 0, int(n/2))

r_2 = np.linspace(0, 1, int(n/2))
g_2 = np.linspace(0.5, 0, int(n/2))
b_2 = np.linspace(0.5, 1, int(n/2))

for idx, a in enumerate(np.linspace(-2, 2, n)):
    v = np.array([1, a])
    Mv = M @ v
    if idx < n/2:
        plt.plot([0, v[0]], [0, v[1]], color=(r_1[idx], g_1[idx], b_1[idx]), linewidth=2, alpha=0.7)
        plt.plot([0, Mv[0]], [0, Mv[1]], color=(r_2[idx], g_2[idx], b_2[idx]), linewidth=2)
    else:
        plt.plot([0, v[0]], [0, v[1]], color=(r_1[n-idx-1], g_1[n-idx-1], b_1[n-idx-1]), linewidth=2, alpha=0.7)
        plt.plot([0, Mv[0]], [0, Mv[1]], color=(r_2[n-idx-1], g_2[n-idx-1], b_2[n-idx-1]), linewidth=2)

plt.axis('off')
plt.axis('square')
plt.gcf().set_size_inches(18, 10)
plt.show()
```
