### __Fermat's Last Theorem__

$$
x^n + y^n = z^n,\newline
\text{has no integer solutions for }n>2
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 100
power=3
x = np.arange(1, num+1)
y = np.arange(1, num+1)

X, Y = np.meshgrid(x, y)

# Cube roots
Fxy = (X**power + Y**power) ** (1/power)

fig, ax = plt.subplots(1, 2, figsize=(12,6))


z_plot = ax[0].imshow(Fxy)


ax[0].set_title('$z = \sqrt[3]{x^3+y^3}$', fontsize=16)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')


# Are there any integers in z
z_is_int = np.zeros((num, num))
for r in range(num):
    for c in range(num):
        if Fxy[r, c]%1 == 0:
            z_is_int[r, c] = 1
        else:
            z_is_int[r, c] = 0

ax[1].imshow(z_is_int, vmin=0, vmax=1)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Is z an integer?', fontsize=16)

print(f'The sum of all elements in matrix: {np.sum(z_is_int)}')

plt.show()
```

```python
num = 100
power=2
x = np.arange(1, num+1)
y = np.arange(1, num+1)

X, Y = np.meshgrid(x, y)

# Cube roots
Fxy = (X**power + Y**power) ** (1/power)

fig, ax = plt.subplots(1, 2, figsize=(12,6))


z_plot = ax[0].imshow(Fxy)
fig.colorbar(z_plot, shrink=.75)

ax[0].set_title('$z = \sqrt[%s]{x^%s+y^%s}$' % (power, power, power), fontsize=16)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')


# Are there any integers in z
z_is_int = Fxy % 1 ==0


z_is_int_plot = ax[1].imshow(z_is_int, vmin=0, vmax=1)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Is z an integer?', fontsize=16)
fig.colorbar(z_is_int_plot, shrink=.75)

print(f'The sum of all elements in matrix: {np.sum(z_is_int)}')

plt.show()
```
