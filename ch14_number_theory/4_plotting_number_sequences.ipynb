### __Number sequences__

Number seq's is an important part of number theory.
We'll discuss how to generate convergent vs. divergent number sequences.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_s = np.arange(1, 81)
 
s1, s2, s3, s4 = np.array([]), np.array([]), np.array([]), np.array([])

for a_idx, a in enumerate(a_s):
    s1 = np.append(s1, a / (a+1))
    s2 = np.append(s2, (a+1) / a)
    s3 = np.append(s3, a / np.sum(a_s[:a_idx+1]))
    s4 = np.append(s4, ((-1)**a) / np.sqrt(a))
   

plt.plot(a_s, s1, label='$\\frac{a}{a+1}$')
plt.plot(a_s, s2, label='$\\frac{a+1}{a}$')
plt.plot(a_s, s3, label='$\\frac{a}{\\sum_{1}^{a}}$')
plt.plot(a_s, s4, label='$\\frac{-1^a}{\sqrt{a}}$')

plt.legend(loc='upper right', fontsize=14)
plt.gcf().set_size_inches(13,8)
plt.show()
```

To understand if a sequence is convergent or divergent we need to examine nominator and denominator and see what grows faster and if a sequence is moving towards some number or not. 
For divergent sequence the nominator should grow faster than the denominator.

```python
a = np.arange(1, 81)
 
s1 = a / (a+1)
s2 = (a+1) / a
s3 = a / np.cumsum(a) #thats' a tricky part, pleast pay attention: cumsum gives a vector of cumulative sums
s4 = ((-1)**a) / np.sqrt(a)
   

plt.plot(a_s, s1, label='$\\frac{a^2}{a+1}$')
plt.plot(a_s, s2, label='$\\frac{a+1}{a}$')
plt.plot(a_s, s3, label='$\\frac{a}{\\sum_{1}^{a}}$')
plt.plot(a_s, s4, label='$\\frac{-1^a}{\sqrt{a}}$')

plt.legend(loc='upper right', fontsize=14)
plt.gcf().set_size_inches(13,8)
plt.show()
```
