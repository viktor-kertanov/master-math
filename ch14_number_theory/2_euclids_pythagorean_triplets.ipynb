### __Euclid's Pythagorean Triplets__

Triplets are integers that satisfy:
    $$a^2 + b^2 = c^2$$
    $$3^2 + 4^2 = 5^2$$
    $$5^2 + 12^2 = 13^2$$

Euclid developed a remarkable formula for finding triplets:

$$

a = m^2 - n^2\newline
b = 2mn\newline
c = m^2 + n^2\newline
m>n\newline
m, n \in\{2, 50\}\newline

$$
   

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euclids_triplets():
    triplets = []
    for m in range(3, 51):
        for n in range(2, 50): #currently we ignoring the constrain m > n, this will give us a "positive" half of the plot
            a = m**2 - n**2
            b = 2*m*n
            c = m**2 + n**2
            # if a > 0 and b > 0 and c > 0:
            triplets.append([a,b,c])
    return triplets

triplets = euclids_triplets()
```

```python
all_a = [t[0] for t in triplets]
all_b  = [t[1] for t in triplets]
all_c = [t[2] for t in triplets]

plt.scatter(all_a, all_b, c=all_c, cmap='viridis', alpha=.5, linewidths=.5, edgecolors='k', s=100)

plt.gcf().set_size_inches(13,9)
plt.axis('off')
plt.show()
```

```python
triplet_eq = [t[0]**2 +t[1]**2 - t[2]**2 for t in triplets]

triplets_eq = np.array(all_a)**2 + np.array(all_b)**2 - np.array(all_c)**2
plt.plot(triplet_eq, 'o')
plt.ylabel('$y=a^2+b^2-c^2$')
# plt.xlim([-1, len(triplet_eq)])

plt.show()
```
