```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np 
from IPython.display import display, Math
import sympy as sym
```

```python
y = np.array([
    [1,1],
    [2,3],
    [3,1 ]
])

display(Math(sym.latex(sym.sympify(y))))
```

```python
y1 = np.array([
    [2,2],
    [2.5, 4],
    [3.5, 1]
])
display(Math(sym.latex(sym.sympify(y1))))
```

```python
p = Polygon(y, facecolor='m', alpha=0.3)
p1 = Polygon(y1, facecolor='None', alpha=1, edgecolor='k', linewidth=0.5)

fig, ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0,4])
ax.set_xlim([0,4])

plt.show()
```
