```python
import sympy as sym
import sympy.plotting.plot as symplot
import numpy as np
```

```python
x, a = sym.symbols('x, a')
fx = a / (x**2 - a)
```

```python
plt = symplot(show=False)
for a_idx, a_val in enumerate(range(1,5)):
    plt.extend(symplot(fx.subs({a: a_val}), show=False))
    plt[a_idx] = '$y = %s / (x^2 - %s)$' % (a_val, a_val)
    plt[a_idx].line_color = list(np.random.rand(3))

plt.xlim = (-5,5)
plt.ylim = (-10,10)
plt.legend = True
plt.show()
```
