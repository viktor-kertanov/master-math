```python
import matplotlib.pyplot as plt
```

```python
p1 = [-3, -1]
p2 = [4, 4]
```

```python
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color = [1, .2, .7], linewidth=3) #color is specifid in a rgb format  

plt.axis('square')
plt.axis([-6,6,-6,6])
plt.show()
```

```python
x = 3
y = 5

plt.plot(x, y, 'ro')
plt.plot([0,x], [0,y], 'r')

plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.plot([-6,6], [0,0], 'k')
plt.plot([0,0], [-6,6], 'k')

plt.show()
```
