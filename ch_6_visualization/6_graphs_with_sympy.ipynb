```python
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x')
y = x**2
y1 = x**3

plt = symplot(y, y1, show=False)
plt.xlim = [-20,20] 
plt.ylim = [-20,20]

plt[0].line_color = 'm'

plt.title = ('This is a nice plot')
plt.show()
```

```python
print(plt)
```

```python
x, a = sym.symbols('x,a')
```

```python
fx = a/x

n=1

plt = symplot(fx.subs({a: n}), (x,-5,5), show=False)
plt2 = symplot(fx.subs({a: n+1}), (x,-5,5), show=False)
plt3 = symplot(fx.subs({a: n+2}), (x,-5,5), show=False)
plt.extend(plt2)
plt.extend(plt3)
plt[2].line_color = 'm'
plt.ylim = (-5,5)


plt[0].label = '$y = %s$' % sym.latex(fx.subs({a:n}))
plt[1].label = '$y = %s$' % sym.latex(fx.subs({a:n+1}))
plt[2].label = '$y = %s$' % sym.latex(fx.subs({a:n+2}))

plt.legend = True

plt.show()
```
