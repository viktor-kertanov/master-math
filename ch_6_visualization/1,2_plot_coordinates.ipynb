```python
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
x = 3
y = 5

plt.plot(x, y, 'bp') #ro -> "red circle", gs-> green square, kp-> black pentagon

plt.axis('square')
plt.axis([-6, 6,-6,6]) #first 2 numbers of X-axis limits, the last 2 numbers go for Y-axis limits
plt.grid()
plt.show()
```

```python
# several points
x = [-4,5,2,4,5,6,3]
y = [0,3,4,5,2,-3,4]

for point in zip(x,y):
    plt.plot(point[0], point[1], 'o', label=f'[x={point[0]},y={point[1]}]')

plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
plt.plot(4, 2, 'r+')

axis = plt.gca()
ylim = axis.get_ylim()
print(f'Our current ylim is: {ylim}')
axis.set_ylim([ylim[0]-3, 10])

plt.show()
```

```python
# exercise
x, y = sym.symbols('x, y')
y = x**2 - 3*x

for x_val in range(-10, 11):
    # print(f'X: {x_val}. Y: {y.subs({x: x_val})}')
    plt.plot(x_val, y.subs({x: x_val}), 'o')

plt.xlabel('x')
plt.ylabel(f'f(x)={y}')

plt.show()
```
