```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_min = -3
x_max = 4
step = 1

x = range(x_min, x_max, step)
y = np.zeros(len(x))
```

```python
y
```

```python
for x_idx, x_val in enumerate(x):
    y[x_idx] = 2 - x_val**2
```

```python
y
```

```python
plt.plot(x,y)

plt.show()
```
