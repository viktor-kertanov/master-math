```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x_min = -3
x_max = 4
step = 1

x = range(x_min, x_max, step)
y = np.zeros(len(x))
```

```python
y
```

```python
for x_idx, x_val in enumerate(x):
    y[x_idx] = 2 - x_val**2
```

```python
y
```

```python
plt.plot(x,y)

plt.show()
```

```python
x = np.linspace(-10,10,100) #lower bound, upper bound and number of steps between points 
print(x)
y1 = 2 + np.sqrt(abs(x))
y2 = 2 - x**2

plt.plot(x,y1, 'm')
plt.plot(x, y2, 'k')

plt.axis('square')
plt.axis([-10,10,-10,10])
plt.show()
```
