# Parabolas

The formula of parabola may look a little bit weird:

$4p(y-k) = (x-h)^2$

If we solve for y, it's getting clearer that h is a shift on x-axis, k is a shift on y-axis, p is some parameter:

$y = \frac{1}{4p} \times (x-h)^2 + k$

#### 1. *Vertex* : (h, k)
#### 2. *Focus*: (h, k+p)
#### 3. *Dirctrix*: y = k - p

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameteres
a = 1 #1/4p
h = 2
k = 1

n = 100

#points
x = np.linspace(h-2, h+2, n)
# directrix = 
y = a * (x-h)**2 + k

plt.plot(x, y)
# plt.plot(y, x) # turning sideways
plt.plot(h, k, 'ro', label='Vertex')
plt.plot(h, k+1/4*a, 'go', label='Focus')

d = k - 1/4*a
plt.plot([x[0], x[-1]], [d,d], label='Directrix')

# plt.axis('square')
# plt.axis([-4, 4, 0, 8])
plt.legend()
plt.grid()
plt.show()
```
