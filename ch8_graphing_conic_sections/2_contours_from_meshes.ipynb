```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a, b = 10, 15

X, Y = np.meshgrid(range(a), range(b))
```

```python
plt.subplot(121) #1 row, 2 columns, current is first plot
plt.pcolormesh(X, edgecolor='k', linewidth = .1)
plt.title("X")

plt.subplot(122)
plt.pcolormesh(Y, edgecolor='k', linewidth= .1)
plt.title("Y")

plt.show()
```

```python
x = np.linspace(-np.pi, 2*np.pi, 40)
y = np.linspace(0, 4*np.pi, 72)

X, Y = np.meshgrid(x, y)
  
Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```

```python
x = np.linspace(-4, 4, 1000)
y = np.linspace(-4, 4, 1000)

sigma_x = 1
sigma_y = 1
mu_x = -2
mu_y = -3

X, Y = np.meshgrid(x, y)

# https://mathworld.wolfram.com/GaussianFunction.html, the formula can be reduces if we have same values for sigma_x = sigma_y = sigmab
Fxy = (1/(2* np.pi * sigma_x * sigma_y)) * np.exp(-( (X - mu_x)**2 / (2*sigma_x**2) + (Y - mu_y)**2 / (2*sigma_y**2) ))

plt.imshow(Fxy)
plt.axis('on')
plt.colorbar()
plt.title('2D Gaussian')
plt.show()
```

Now I'll use the formula from the video:
$ Z = \exp^{-\frac{(X^2+Y^2)}{s}}$

```python
x = np.linspace(-10, 10, 1000)
y = np.linspace(-10, 10, 1000)

X, Y = np.meshgrid(x, y)

sigma = 40

Fxy = np.exp(-(X**2 + Y**2)/sigma)

plt.imshow(Fxy)

plt.show()
```
