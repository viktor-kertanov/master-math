```python
import  numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = .5
k = 2
h = 1

axlim = 2*np.max([a, b]) + np.max([abs(k), abs(h)])
x = np.linspace(-axlim, axlim, 100)

X, Y = np.meshgrid(x, x)

Fxy = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1
```

```python
plt.contour(X, Y, Fxy, 0)
plt.plot(h, k, 'go')
plt.plot([-axlim, axlim], [0, 0], 'k--', linewidth=0.5, alpha=0.5)
plt.plot([0, 0], [-axlim, axlim],  'k--', linewidth=0.5, alpha=0.5)
plt.gca().set_aspect('equal')
plt.show()
```
