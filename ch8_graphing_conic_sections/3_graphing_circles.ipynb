# Graphing Circles

The general formula for drawing a circle is:

$(y-k)^2 + (x-h)^2 = r^2$

In school it usually looks like this:

$y^2 + x^2 = r^2$

There are, actually, a few ways to draw a circle. Previously we used a unit-circle with sine and cosine functions. And here we'll use another method.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parameters
h = 2
k = -3
r = 3   

axlim = r + np.max([abs(h), abs(k)])

x = np.linspace(-axlim, axlim, 100)

X, Y = np.meshgrid(x, x)
```

```python
plt.subplot(121)
plt.imshow(X)

plt.subplot(122)
plt.imshow(Y)
plt.axis('off')

plt.show()
```

```python
Fxy = (X - h)**2 + (Y - k)**2 - r**2

plt.contour(X, Y, Fxy, 0)

plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'k--')
plt.plot(h, k, 'ro')


plt.gca().set_aspect('equal') #cool method
plt.show()
```

```python
# Exercise
h_s = [-1.5, 1.5]
k = 0

radii = np.linspace(.5, 3, 30)
n = 100

for h in h_s:
    for r_idx, r in enumerate(radii):
        axlim = r + np.max([abs(h), abs(k)])
        x = np.linspace(-axlim, axlim, n)
        X, Y = np.meshgrid(x, x)

        Fxy = (X-h)**2 + (Y-k)**2 - r**2

        grayness= 0.6
        gray_color = grayness+ (1-grayness)*((r_idx)/len(radii))
        plt.contour(X, Y, Fxy, 0, colors=(gray_color, gray_color, gray_color))

plt.plot([-1.5, 1.5], [k, k], 'k', linewidth=4)
plt.gca().set_aspect('equal')

plt.axis('off')
plt.show()

```

```python
# Exercise
h_s = [-1.5, 1.5]
k = 0

radii = np.linspace(.05, 3, 100)
n = 1000

for h in h_s:
    for r_idx, r in enumerate(radii):
        axlim = r + np.max([abs(h), abs(k)])+1
        x = np.linspace(-axlim, axlim, n)
        X, Y = np.meshgrid(x, x)

        Fxy = (X-h)**2 + (Y-k)**2 - r**2

        plt.contour(X, Y, Fxy, 0, colors='k', linewidths=.1)

plt.gca().set_aspect('equal')

plt.axis('off')
plt.savefig('circles_axi.svg')
plt.show()
```
