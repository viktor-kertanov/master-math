# Graphing Ellipses

- Ellipses has 2 radii (in a circle they are equal and = r, in an ellipse these are a and b).
- "b" is a radius by which the ellipse is elevated on the y-axis. 
- "a" is a radius by which the ellipse is shifted along the x-axis.

**Notabene**: it is clear that if a = b, then we are getting back to our circle equation.

$\frac{(y-k)^2}{b^2} + \frac{(x-h)^2}{a^2} - 1 = 0  $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0
k = 0
a = 5
b = 2

axlim = np.max([a, b]) + np.max([abs(h), abs(k)]) + 1
```

```python
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

Fxy = (X - h)**2/a**2 + (Y-k)**2/b**2 - 1

plt.contour(X, Y, Fxy, 0, colors='red')

plt.plot([-axlim, axlim], [0, 0], 'k--', alpha=0.8, linewidth=0.5)
plt.plot([0, 0], [-axlim, axlim],  'k--', alpha=0.8, linewidth=0.5)
plt.gca().set_aspect('equal')

plt.show()
```

```python
# Exercise: from a circle to a few ellipses
b = 4
h = 0

n = 100

a_points_n = 16

for a_idx, a in enumerate(np.linspace(-4, 4, a_points_n), start=1):
    axlim = np.max([np.max(a), b]) + np.max([abs(h), abs(np.max(k))])+1
    x = np.linspace(-axlim, axlim, n)
    X, Y = np.meshgrid(x, x)

    # print(f"{a_idx}. {a}")
    Fxy = (X - h)**2 / a**2 + (Y - a)**2 / b**2 - 1
    if a > 0:
        plt.contour(X, Y, Fxy, 0, colors=(abs(a)/4, abs(a)/10, abs(a)/4 ))
    else:
        color_scale= 0.01
        color = color_scale+(1-color_scale)*abs(a)/a_points_n
        plt.contour(X, Y, Fxy, 0, colors=(color, color, color))


plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```

```python
# Extension to the exercise
# Exercise: from a circle to a few ellipses
b = 4
h = 0

n = 100

a_points_n = 32

for a_idx, a in enumerate(np.linspace(-4, 4, a_points_n), start=1):
    for h in np.linspace(-5, 5, 1):
        for b in np.linspace(-4, 4, 4):
            axlim = np.max([np.max(a), b]) + np.max([abs(h), abs(np.max(k))])+3
            x = np.linspace(-axlim, axlim, n)
            X, Y = np.meshgrid(x, x)

            # print(f"{a_idx}. {a}")
            Fxy = (X - h)**2 / a**2 + (Y - a)**2 / b**2 - 1
            if a > 0:
                plt.contour(X, Y, Fxy, 0, colors='k', linewidths=.5)
            else:
                color_scale= 0.01
                color = color_scale+(1-color_scale)*abs(a)/a_points_n
                plt.contour(X, Y, Fxy, 0, colors='k', linewidths=.5)


plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('ellipses_self_amuse.svg')
plt.show()
```
