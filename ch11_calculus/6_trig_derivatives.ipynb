Imagine the deriviatives of sine and cosine as a circle.

```python
x = sym.symbols('x')

fx = [sym.sin(x), sym.cos(x), -sym.sin(x), -sym.cos(x)]

for f in fx:
    display(Math('f: %s \\quad \\Rightarrow \\quad f\': %s' % (sym.latex(f), sym.latex(sym.diff(f)))))
```

```python
# Another way, from the course

x = sym.symbols('x')
fx = sym.sin(x)

for _ in range(8):
    display(Math('\\frac{df}{dx} %s = %s' % (sym.latex(fx), sym.latex(sym.diff(fx)))))
    fx = sym.diff(fx)
```

```python
# Now let's plot all this stuff
x = sym.symbols('x')
fx = sym.sin(x)

for i in range(4):
    if i == 0:
        plt = symplot(fx, show=False, label=f'{i}. {fx}', line_color=(i/5, i/3, i/5))
        fx = sym.diff(fx)
        continue
    
    plt.extend(symplot(fx, show=False, label=f'{i}. {fx}', line_color=(i/5, i/3, i/5)))
    
    fx = sym.diff(fx)

plt.legend = True
plt.xlim = [-3,3]
plt.show()
```
