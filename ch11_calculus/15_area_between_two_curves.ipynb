### **Area between two curves**

If we have two curves, then we can calculate the area between them and that's pretty straightforward: <br>

$$ \int_{a}^{b} [f(x)-g(x)] dx$$

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx_symbol = x**2
gx_symbol = x

fx = sym.lambdify(x, fx_symbol)
gx = sym.lambdify(x, gx_symbol)

x_val = np.linspace(-2, 2, 1001)

# add patch
x_patch = np.linspace(0, 1, 1001)
y_patch = np.vstack([fx(x_patch), gx(x_patch)])


plt.plot(x_val, fx(x_val))
plt.plot(x_val, gx(x_val), 'r')

plt.legend(['$f(x) = %s$' % sym.latex(fx_symbol), '$g(x) = %s$' % sym.latex(gx_symbol)])
plt.axis([-.25, 1.25, -.5, 1.5]) #zooming in the area in which we are interested in

plt.show()
```
