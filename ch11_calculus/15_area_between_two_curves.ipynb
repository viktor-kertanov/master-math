### **Area between two curves**

If we have two curves, then we can calculate the area between them and that's pretty straightforward: <br>

$$ \int_{a}^{b} [f(x)-g(x)] dx$$

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx_symbol = x**2
gx_symbol = x

fx = sym.lambdify(x, fx_symbol)
gx = sym.lambdify(x, gx_symbol)

x_val = np.linspace(-2, 2, 1001)

# add patch
x_patch = np.linspace(0, 1, 1001)
y_patch = np.vstack([gx(x_patch), fx(x_patch)]).T 

fig, ax = plt.subplots()
ax.add_patch(Polygon(y_patch, facecolor='k', alpha=0.3))

plt.plot(x_val, fx(x_val), 'b', label='$f(x) = %s$' % sym.latex(fx_symbol))
plt.plot(x_val, gx(x_val), 'r', label='$g(x) = %s$' % sym.latex(gx_symbol))

plt.legend()
plt.axis([-.25, 1.25, -.5, 1.5]) #zooming in the area in which we are interested in

plt.show()
```
