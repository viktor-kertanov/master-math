```python
import sympy.plotting.plot as symplot
from IPython.display import display, Math
import sympy as sym
```

```python
x, a = sym.symbols('x, a')
a_list = [0, 1, 2, 3]

colors = 'brkm'

for a in a_list:
    fx = sym.cos(x + sym.sin(x)) + a
    if a == 0:
        plt = symplot(fx, show=False, label=fx, line_color=colors[a])
    else:
        plt.extend(symplot(fx, show=False, label=fx, line_color=colors[a]))

plt.title = 'The functions'
plt.legend = True
plt.show()
```

```python
# Now let's plot the derivatives and not the original functions
# The derivatives look the same because the functions behave the same: grow the same, decline the same etc. a parameter does not effect the behaviour  

x, a = sym.symbols('x, a')
a_list = [0, 1, 2, 3]

for a in a_list:
    diff_fx = sym.diff(sym.cos(x + sym.sin(x)) + a)
    if a == 0:
        plt = symplot(diff_fx, show=False, label=diff_fx)
    else:
        plt.extend(symplot(diff_fx, show=False, label=diff_fx))

plt.title = 'The derivatives'
plt.legend = True
plt.show()
```
