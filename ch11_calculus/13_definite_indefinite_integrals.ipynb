### **Definite and Indefinite integrals**

The concept of an integral is simple: adding together a lot of really tiny, tiny things. The area between two points:
<br>

$$\int_{a}^{b} f(x) dx \newline$$
<br>

And if we look at this integral symbol, we'll see:
1. Boundaries **a** and **b** (on x-axis);
2. __f(x)__ &ndash; that's actually the __height__ of the "bar";
3. __dx__ &ndash; that's the discretization, the __smallest x we can obtain__.
4. __f(x)*dx__ &ndash; is the __area of one "bar"__
5. $\int$ is actually the sum of all the bars, __total area__

<br>

If we have __a__ and __b__ parameters than we are dealing with a definite integral and we want to calculate exactly the area under the integral.
If we don't have __a__ and __b__ then we are dealing with an indefinite integral.

Indefinite integrals may be very hard to calculate, as you have to know a lot of techniques and tricks.
Though integrating polynomials is pretty straightforward: <br>

$$\int bx^a dx = \frac{bx^{a+1}}{a+1} +c $$ 


$$\int 4x^2 dx = \frac{4x^{3}}{3} +c $$ 

<br>
<br>
And if we take a derivative from the right part of the equation then we'll have: <br>

$$\left( \frac{bx^{a+1}}{a+1} + c \right)'= \frac{b(a+1)x^{a+1-1}}{a+1} + 0 = b*x^{a}$$

```python
import sympy as sym
from IPython.display import display, Math
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x')
f = x

sym.integrate(f) #indefinite integral
```

```python
# definite integral

sym.integrate(f, (x, 0, 1)) 
# the area is under a line which goes by 45% up, so there's a triangle with sides 1,1 and hypothenusa. it's area is 1/2 -- half of a square with sides 1/1
```

```python
p = symplot(f, show=False)
p.xlim = [-1, 3]
p.ylim = [-1, 3]
p.show()
```

```python
f = x**3 / (x-2)
integral = sym.integrate(f)

p = symplot(f, show=False)
p.extend(symplot(integral, (x, 2.1, 10), show=False, line_color='r'))

p[0].label = '$f(x) = %s$' % sym.latex(f)
p[1].label = '$\\int f(x)dx = %s$' % sym.latex(integral)

p.ylim = [-200, 200]
p.legend = True
p.show( )
```
