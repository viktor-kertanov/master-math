# Derivatives

A huge topic and it is possible to spend a whole semester just exploring the derivatives.
Here we'll cover some basics in connection with Python.

In case of a derivative we care not about the function itself and how it looks on a graph, but about how the function changes along the way (is it going down/up, extrema etc.)
If a function is going down -- slowly -- the derivative will be negative, but small. If it is going down fast, the derivative will be negative and big.

Often a derivative is written down like this:

$\frac{df}{dx}x^2=2x^1\newline$<br>
$\frac{d}{dx}x^3=3x^2 \newline$<br>
$\frac{d}{dx}(ax^n) = nax^{n-1} \newline$<br>

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x')
fx = x**2
dfx = sym.diff(fx)

dfx
```

```python
# Leibniz notation
display(Math('f(x) = %s, \\quad \\frac{df}{dx}=%s' % (sym.latex(fx), sym.latex(dfx))))
```

```python
# Lagrange notation, pronounced "f-prime"
display(Math('f(x) = %s, \\quad f\'=%s' % (sym.latex(fx), sym.latex(dfx))))
```

```python
# Newton notation
display(Math('f(x) = %s, \\quad \\dot{f}=%s' % (sym.latex(fx), sym.latex(dfx))))
```

```python
# symplotting

fx = 3 - x**3

p = symplot(fx, (x, -3, 3), show=False)

p.extend(symplot(sym.diff(fx), show=False))
p[1].line_color = 'r'
p[0].label = '$f(x) = %s$' % sym.latex(fx)
p[1].label = '$f(x) = %s$' % sym.latex(sym.diff(fx))

p.legend=True
p.ylim = [-10, 10]
p.xlim = [-3, 3]
p.show( )
```
