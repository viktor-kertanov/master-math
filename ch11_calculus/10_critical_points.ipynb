# Critical points

Critical points are the points where the derivative is equal to zero (i.e. df = 0, function "stopped" while dx > 0 [it cannot be zero as it is in the denominator]).
Also derivative may be undefined in critical points (that's a different case)

```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from IPython.display import display, Math
```

```python
# emperical method

x = np.linspace(-5, 5, 1001) # something and plus something should be in odd number of steps, as we have zero in the middle
fx = x**2 * np.exp(-x**2)

dfx = np.diff(fx)/np.diff(x) #it's just descrete  derivative, calculated in each point, not a symbolic thing like in sympy

max_peaks = find_peaks(fx)[0]
min_peaks = find_peaks(-fx)[0]

plt.plot(x, fx)
plt.plot(x[max_peaks], fx[max_peaks], 'go')
plt.plot(x[min_peaks], fx[min_peaks], 'ro')
plt.plot(x[:-1], dfx)
plt.plot([-5, 5], [0, 0], 'k--', alpha=0.5, linewidth=0.5)
plt.plot([0, 0], [-1, 1], 'k--', alpha=0.5, linewidth=0.5)

# plt.axis('square')

# plt.axis([-1.5, 1.5, -1, 3])
plt.show()
```

```python
# now a symbolic method
x = sym.symbols('x')
fx = x**2 * sym.exp(-x**2)

dfx = sym.diff(fx)
display(Math(sym.latex(dfx)))
critical_points = sym.solve(dfx)
```
