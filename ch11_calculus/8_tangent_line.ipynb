# Tangent Line (Касательная)

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
x = sym.symbols('x')

#fx definition and its derivative
f = x**2
diff_f = sym.diff(f)

#value x at which to compute the tangent line
x_a = 1

#get the function and derivative value at point 'a'
f_a = f.subs(x, x_a)
df_fa = diff_f.subs(x, x_a)
print(f_a)
print(df_fa)

#converting to numpy object
x_np = np.linspace(-2, 2, 200)
f_np = sym.lambdify(x, f)(x_np)
diff_f_np = sym.lambdify(x, diff_f)(x_np)
 
#compute the tangent line
tangent_line = df_fa * (x_np - x_a) + f_a

plt.plot(x_np, f_np, label='f(x)')
plt.plot(x_np, diff_f_np, label='f(x)\'')
plt.plot(x_np, tangent_line, label=f'tangent at a={x_a}')
plt.plot(x_a, f_a, 'ro')


plt.axis('square')
plt.axis([-2, 2, -.5, 2])

ax = plt.gca() #get the info about the current axis
plt.plot(ax.get_xlim(), [0,0], 'k--', alpha=0.5)
plt.plot([0,0], ax.get_ylim(),  'k--', alpha=0.5)
plt.legend()
plt.show()
```
