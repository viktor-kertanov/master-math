```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
```

```python
# Piecewise function, кусочная функция

piece_1 = 0
piece_2 = -2*x
piece_3 = x**3/10

fx = sym.Piecewise(
    (piece_1, x < 0),
    (piece_2, (x>=0) & (x < 3)),
    (piece_3, x >= 3)
)

fx
```

```python
fx_lambda = sym.lambdify(x, fx)
```

```python
x_values = np.linspace(-3, 5, 2000)

plt.plot(x_values, fx_lambda(x_values), 'r')

plt.show()
```

```python
# Exercise: piecewise function
x = sym.symbols('x')

p_1 = x ** 3
p_2 = sym.log(x, 2)

gx = sym.Piecewise(
    (p_1, x <=0 ),
    (p_2, x > 0)
)

gx_lambda = sym.lambdify(x, gx)

gx
```

```python
x_val = np.linspace(-2, 2, 1000)

with plt.xkcd():
    plt.plot(x_val, gx_lambda(x_val), 'k')
    plt.ylabel('y')
    plt.xlabel('x')
    plt.show()
```
