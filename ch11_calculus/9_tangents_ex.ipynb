```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
x = sym.symbols('x')
fx = x**2
```

```python
def get_tangent_line(f, x_a: float, bounds: list):
    diff_f = sym.diff(f)
    f_a = fx.subs(x, x_a)
    diff_f_a = diff_f.subs(x, x_a)

    x_np = np.linspace(bounds[0], bounds[1], 200)
    f_np = sym.lambdify(x, f)(x_np)
    diff_f_np = sym.lambdify(x, diff_f)(x_np)

    return diff_f_a * (x_np - x_a) + f_a
```

```python
x_bounds = [-2, 2]
x_np = np.linspace(x_bounds[0], x_bounds[1], 200)
fx_np = sym.lambdify(x, fx)(x_np)
plt.plot(x_np, fx_np, 'k', linewidth=5)

for a in x_np:
    tangent_line = get_tangent_line(fx, a, x_bounds)
    plt.plot(x_np, tangent_line, color=(abs(a)/3, abs(a)/4, abs(a)/2), linewidth=1)

plt.axis('square')  
plt.axis('off')
plt.axis([-2, 2, -1, 3])
plt.show()
```
