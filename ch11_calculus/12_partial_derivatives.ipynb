### **Partial derivatives**

Partial derivatives are useful when we have a function of more than one variable, e.g.:
$f(x,y) = 2xy^2\newline$ <br>
We need to understand against which variable we calculate the derivative.
__The partial derivative of f with respect to x__ or __partial f, partial x__

$\frac{\partial f}{\partial x} = f_x = 2y^2 \newline$<br><br>
$\frac{\partial f}{\partial y} = f_y = 4xy$

```python
import sympy as sym
from IPython.display import display, Math
from sympy.plotting import plot3d_parametric_surface as symplot3d
import sympy.plotting.plot as symplot
```

```python
x, y = sym.symbols('x, y')

f_xy = x**2 + x*y**2

display(Math('\\frac{\\partial f}{\\partial x} = %s' % sym.latex(sym.diff(f_xy, x))))
display(Math('\\frac{\\partial f}{\\partial y} = %s' % sym.latex(sym.diff(f_xy, y))))
```

```python
# exercise
f_xy_plot = symplot3d(x, y, f_xy, show=False)
df_x_plot = symplot3d(x, y, sym.diff(f_xy, x), show=False)
df_y_plot = symplot3d(x, y, sym.diff(f_xy, y), show=False)

f_xy_plot.title = '$f(x, y)=%s$'%sym.latex(f_xy)
df_x_plot.title = '$f_x=%s$'%sym.latex(sym.diff(f_xy, x))
df_y_plot.title = '$f_y=%s$'%sym.latex(sym.diff(f_xy, y))
f_xy_plot.show()
df_x_plot.show()
df_y_plot.show()
```

```python
# Solution from the course

p = sym.plotting.plot3d(f_xy, (x, -3, 3), (y, -3, 3), title='$f(x,y)=%s$'%sym.latex(f_xy))
p = sym.plotting.plot3d(sym.diff(f_xy, x), (x, -3, 3), (y, -3, 3), title='$f(x,y)=%s$'%sym.latex(sym.diff(f_xy, x)))
p = sym.plotting.plot3d(sym.diff(f_xy, y), (x, -3, 3), (y, -3, 3), title='$f(x,y)=%s$'%sym.latex(sym.diff(f_xy, y)))
```
