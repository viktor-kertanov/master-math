```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Math, display
```

```python
x = sym.symbols('x')

fx = sym.exp(-x)*sym.sqrt(x+1)
fx_lambda = sym.lambdify(x, fx)
print(fx_lambda(10))

gx = sym.cos(x+sym.sin(x))
gx_lambda = sym.lambdify(x, gx)
print(gx_lambda(10))
```

```python
x_val = np.linspace(-10, 10, 1000)

plt.plot(x_val, fx_lambda(x_val), 'r', linewidth=1)
plt.show()

plt.plot(x_val, gx_lambda(x_val), 'g', linewidth=1)
plt.show()
```

```python
limit_x = 5
limit_fx = sym.limit(fx, x, limit_x)
limit_gx = sym.limit(gx, x, limit_x)
limit_division_x = sym.limit(fx/gx, x, limit_x)

display(Math('\\text{Lets prove that: }\\frac{\\lim_{x \\to %g}f(x)}{\\lim_{x \\to %g}g(x)} = \\lim_{x \\to %g} \\frac{f(x)}{g(x)}' % (limit_x, limit_x, limit_x)))
display(Math('\\text{We have:}\\\  \\frac{%s} {%s} = %s' % (sym.latex(limit_fx), sym.latex(limit_gx), sym.latex(limit_division_x))))
display(Math('\\text{Thats equal to:} \\newline %s = %s' % (sym.latex(limit_fx/limit_gx), sym.latex(limit_division_x))))
display(Math('\\text{Thats equal to:} \\newline %s = %s' % ((limit_fx/limit_gx).evalf(), limit_division_x.evalf())))
```
