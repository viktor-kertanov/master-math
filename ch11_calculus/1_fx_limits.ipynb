# Limit of a function

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Math, display
```

```python
x = sym.symbols('x')
fx = x**3

limit_point = 1.5
limit = sym.limit(fx, x, limit_point)

display(Math('\\lim_{x\\to%g} %s = %g' % (limit_point, sym.latex(fx), limit)))
```

We can use lambdify function that creates an output (also a function) and we can change x-variable to get different results of fx. <br>
Similar to what we do in numpy.

```python
fx_calc = sym.lambdify(x, fx)
x_val = np.linspace(-5, 5, 200)

plt.plot(x_val, fx_calc(x_val), 'b')
plt.plot(limit_point, limit, 'ro')

# plt.axis('square')
# plt.axis([-5, 5, -100, 100])
plt.show()
```

```python
fx_2 = x**2 / (x-2)
fx_2_calc = sym.lambdify(x, fx_2)

x_val_2 = np.linspace(1, 3, 1000)

limit_x = 2
limit_y = sym.limit(fx_2, x, limit_x, dir='-')
display(Math('\\lim_{x\\to%g^{-}} %s = %g' % (limit_x, sym.latex(fx_2), limit_y)))

plt.plot(x_val_2, fx_2_calc(x_val_2))
plt.plot(limit_x, limit_y, 'go')
plt.show()
```
